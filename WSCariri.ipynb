4. **Baixar Arquivos ZIP, Extrair CSVs e Salvar Individualmente (com verifica√ß√£o)**

````python
# Importar tqdm para a barra de progresso (adicione no in√≠cio do seu script se ainda n√£o estiver l√°)
from tqdm.notebook import tqdm

# Etapa 4: Baixar Arquivos ZIP, Extrair CSVs e Salvar Individualmente (com verifica√ß√£o)
print('=' * 80)
print("--- Iniciando Etapa 4: Download e salvamento de arquivos CSV individuais ---")
print('=' * 80)

# Adicionando a barra de progresso com tqdm
for url in tqdm(urls_dados, desc="Processando arquivos da Rede SONDA"):
    nome_arquivo_origem = url.split('/')[-1]
    csv_filename_destino = nome_arquivo_origem.replace('.zip', '.csv').replace('.7z', '.csv') # Adicionado .replace('.7z', '.csv') para consist√™ncia com 2009
    csv_path_destino = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, csv_filename_destino)

    # Verificar se o arquivo CSV individual j√° existe
    if os.path.exists(csv_path_destino):
        try:
            df_existente_check = pd.read_csv(csv_path_destino, sep=";", header=None, nrows=1, encoding="latin1")
            if df_existente_check.shape[1] != len(colunas):
                print(f"üîÑ Arquivo '{csv_filename_destino}' com formato inv√°lido. Removendo e reprocessando.")
                logging.warning(f"Arquivo '{csv_filename_destino}' ({df_existente_check.shape[1]} cols) com formato de colunas diferente do esperado ({len(colunas)} cols). Removendo para reprocessamento.")
                os.remove(csv_path_destino)
            else:
                print(f"‚è© Arquivo '{csv_filename_destino}' j√° existe e tem formato correto. Pulando download.")
                logging.info(f"Arquivo '{csv_filename_destino}' j√° existe e tem o formato correto ({len(colunas)} cols). Pulando download e processamento.")
                continue
        except pd.errors.EmptyDataError:
            print(f"üóëÔ∏è Arquivo '{csv_filename_destino}' existe mas est√° vazio. Removendo para reprocessar.")
            logging.warning(f"Arquivo '{csv_filename_destino}' existe mas est√° vazio. Removendo para reprocessamento.")
            os.remove(csv_path_destino)
        except Exception as e:
            print(f"‚ùå Erro ao verificar '{csv_filename_destino}': {e}. Removendo para reprocessar.")
            logging.error(f"Erro ao verificar arquivo existente '{csv_filename_destino}': {e}. Removendo para tentar reprocessamento.")
            try:
                os.remove(csv_path_destino)
            except OSError as oe:
                print(f"N√£o foi poss√≠vel remover '{csv_filename_destino}': {oe}")
                logging.error(f"N√£o foi poss√≠vel remover o arquivo '{csv_filename_destino}' para reprocessamento: {oe}")

    try:
        # print(f"Baixando e processando '{url}'...") # Pode manter ou remover, dependendo do qu√£o detalhado quer o output.
        logging.info(f"Baixando e processando '{url}'...")
        response = requests.get(url, timeout=30) # Adicionado timeout para evitar travamentos indefinidos
        response.raise_for_status() # Levanta um erro HTTP para c√≥digos de status ruins (4XX ou 5XX)

        extensao_arquivo = nome_arquivo_origem.split('.')[-1].lower()
        df_processado = None

        if extensao_arquivo == 'zip':
            # Abrir o arquivo ZIP a partir do conte√∫do baixado em mem√≥ria
            with zipfile.ZipFile(BytesIO(response.content)) as z:
                # Listar todos os arquivos dentro do ZIP
                nomes_csv_no_zip = [f_name for f_name in z.namelist() if f_name.lower().endswith(".csv")] # Considerar .csv em min√∫sculas
                if not nomes_csv_no_zip:
                    print(f"‚ö†Ô∏è Nenhum arquivo .csv encontrado no ZIP: {url}")
                    logging.warning(f"Nenhum arquivo .csv encontrado dentro do ZIP: {url}")
                    continue
                nome_csv_principal_zip = nomes_csv_no_zip[0]
                logging.info(f"Encontrado arquivo CSV '{nome_csv_principal_zip}' dentro do ZIP.")
                # Abrir o arquivo CSV de dentro do ZIP
                with z.open(nome_csv_principal_zip) as f_csv:
                    temp_df = pd.read_csv(f_csv, sep=";", header=None, encoding="latin1", low_memory=False)
                    # Certifique-se de que a atribui√ß√£o a df_processado ocorra aqui para o bloco de checagem de colunas
                    df_processado = temp_df

        elif extensao_arquivo == '7z':
            temp_extract_dir = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, f"temp_extract_{nome_arquivo_origem}")
            os.makedirs(temp_extract_dir, exist_ok=True)
            logging.info(f"Criado diret√≥rio tempor√°rio para 7z: {temp_extract_dir}")
            try:
                with py7zr.SevenZipFile(BytesIO(response.content), mode='r') as archive:
                    nomes_arquivos_no_7z = archive.getnames()
                    nomes_csv_no_7z = [f_name for f_name in nomes_arquivos_no_7z if f_name.lower().endswith('.csv')]
                    if not nomes_csv_no_7z:
                        print(f"‚ö†Ô∏è Nenhum arquivo .csv encontrado dentro do 7z: {url}")
                        logging.warning(f"Nenhum arquivo .csv encontrado dentro do 7z: {url}")
                        continue
                    nome_csv_principal_7z = nomes_csv_no_7z[0]
                    logging.info(f"Extraindo '{nome_csv_principal_7z}' de dentro do 7z '{nome_arquivo_origem}' para {temp_extract_dir}")
                    archive.extract(path=temp_extract_dir, targets=[nome_csv_principal_7z])
                    caminho_csv_extraido_temp = os.path.join(temp_extract_dir, nome_csv_principal_7z)
                    temp_df = pd.read_csv(caminho_csv_extraido_temp, sep=";", header=None, encoding="latin1", low_memory=False)
                    # Certifique-se de que a atribui√ß√£o a df_processado ocorra aqui
                    df_processado = temp_df
            finally:
                if os.path.exists(temp_extract_dir):
                    shutil.rmtree(temp_extract_dir)
                    logging.info(f"Diret√≥rio tempor√°rio {temp_extract_dir} removido.")
        else:
            print(f"‚ö†Ô∏è Extens√£o de arquivo n√£o suportada '{extensao_arquivo}' para {url}")
            logging.warning(f"Extens√£o de arquivo n√£o suportada '{extensao_arquivo}' para {url}")
            continue

        if df_processado is not None:
            # Corre√ß√£o para o caso de arquivos de 2009 com 12 colunas, remover a √∫ltima
            # Esta l√≥gica deve ser aplicada ao df_processado, que √© o que cont√©m os dados lidos
            if "09" in csv_filename_destino and df_processado.shape[1] == 12:
                df_processado = df_processado.iloc[:, :-1]
                print(f"‚öôÔ∏è Coluna extra removida de '{csv_filename_destino}' (original: {nome_arquivo_origem}).")
                logging.info(f"√öltima coluna removida do arquivo '{csv_filename_destino}' (original: {nome_arquivo_origem}), novo shape: {df_processado.shape}")

            if df_processado.shape[1] == len(colunas):
                df_processado.to_csv(target_csv_path, index=False, header=False, sep=";", encoding="latin1")
                print(f"‚úÖ Arquivo salvo: '{csv_filename_destino}' em '{target_csv_path}'.") # Mensagem de sucesso com caminho
                logging.info(f"Arquivo processado e salvo como: '{target_csv_path}' com {df_processado.shape[0]} linhas e {df_processado.shape[1]} colunas.")
            else:
                print(f"‚ùå ERRO: '{csv_filename_destino}' (de {nome_arquivo_origem}) tem {df_processado.shape[1]} cols, esperado {len(colunas)}. N√ÉO SALVO.")
                logging.error(f"Ap√≥s processamento, arquivo '{csv_filename_destino}' (de {nome_arquivo_origem}) possui {df_processado.shape[1]} colunas, mas o esperado era {len(colunas)}. Arquivo n√£o salvo.")
        else:
            print(f"‚ö†Ô∏è Nenhum DataFrame foi processado para a URL {url}")
            logging.warning(f"Nenhum DataFrame foi processado para a URL {url}.")

    except requests.exceptions.RequestException as e:
        print(f"‚ùå Erro de requisi√ß√£o ao baixar {url}: {e}")
        logging.error(f"Erro de requisi√ß√£o ao baixar {url}: {e}")
    except (zipfile.BadZipFile, py7zr.exceptions.Bad7zFile) as e:
        print(f"‚ùå Erro: Arquivo baixado de {url} corrompido ou inv√°lido.")
        logging.error(f"Erro: Arquivo baixado de {url} (ZIP/7Z) n√£o √© v√°lido ou est√° corrompido: {e}")
    except Exception as e:
        print(f"‚ùå ERRO INESPERADO ao processar {url}: {str(e)}")
        logging.error(f"Erro inesperado ao processar {url}: {str(e)}")

print("-" * 80)
print("Etapa 4 conclu√≠da!!!")
print("-" * 80)
````

````python
# --- ETAPA 1: DEFINIR CAMINHOS DE DESTINO ---
import os
import pandas as pd
import requests
import zipfile
import shutil
from io import BytesIO
import logging
from IPython.display import display # Para exibir DataFrames de forma mais elegante no Colab
from tqdm.notebook import tqdm # Importando tqdm para barras de progresso no Jupyter/Colab

# Configurar logging (√∫til para um rastreamento mais detalhado se necess√°rio)
# Se esta c√©lula for executada m√∫ltiplas vezes, basicConfig pode n√£o ter efeito ap√≥s a primeira vez
# sem force=True (Python 3.8+), mas as chamadas logging.info() ainda funcionar√£o.
if not logging.getLogger().handlers: # Configura apenas se nenhum handler estiver configurado
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Instala√ß√£o autom√°tica de py7zr, se necess√°rio
try:
    import py7zr
    logging.info("py7zr j√° est√° instalado.") # Mensagem de log, n√£o print
except ImportError:
    logging.info("py7zr n√£o encontrado. Tentando instalar...") # Mensagem de log
    import subprocess
    import sys
    try:
        print("Instalando py7zr... Aguarde.") # Mensagem para o usu√°rio
        subprocess.check_call([sys.executable, "-m", "pip", "install", "py7zr"])
        import py7zr
        print("py7zr instalado com sucesso.")
        logging.info("py7zr instalado com sucesso via pip.")
    except Exception as e:
        print(f"ERRO: Falha ao instalar py7zr: {e}")
        logging.error(f"Falha ao instalar py7zr: {e}")
        raise ImportError("py7zr n√£o p√¥de ser instalado. O script n√£o pode continuar sem ele para arquivos .7z.")

print('=' *80)
print("--- Iniciando a ETAPA 1: Defini√ß√µes Iniciais e Configura√ß√µes ---")
print('=' *80)

# Configura√ß√µes globais
BASE_DRIVE_PATH = '/content/drive/MyDrive/Disciplinas/CieÃÇncia de Dados/Projeto de Disciplina' # Ajuste se necess√°rio
HEADER_FILE_PATH = os.path.join(BASE_DRIVE_PATH, 'WD_header.csv')
INDIVIDUAL_CSV_SAVE_PATH = os.path.join(BASE_DRIVE_PATH, 'DadosSonda_Individuais_Jun2007Set2009')
FINAL_DATASET_SAVE_PATH = os.path.join(BASE_DRIVE_PATH, 'DadosSonda_Unificados_Jun2007Set2009')
FINAL_DATASET_NAME = 'dfSonda_jun2007_set2009.csv'

# Montar Google Drive (se ainda n√£o montado) - geralmente feito em c√©lula separada
# from google.colab import drive
# drive.mount('/content/drive')
# print("\nGoogle Drive Montado (ou j√° estava montado).") # Adicionar ap√≥s a montagem real

os.makedirs(INDIVIDUAL_CSV_SAVE_PATH, exist_ok=True)
os.makedirs(FINAL_DATASET_SAVE_PATH, exist_ok=True)

print(f"\nCaminho do arquivo de cabe√ßalho: \n{HEADER_FILE_PATH}")
print(f"\nPasta para CSVs individuais: \n{INDIVIDUAL_CSV_SAVE_PATH}")
print(f"\nPasta para dataset final: \n{FINAL_DATASET_SAVE_PATH}")
print('-' *80)
print("ETAPA 1 Conclu√≠da: Caminhos definidos e pastas criadas/verificadas.\n")

# --- ETAPA 2: CARREGAR CABE√áALHO DE REFER√äNCIA ---
print('=' * 80)
print("--- Iniciando a ETAPA 2: Carregar Cabe√ßalho de Refer√™ncia ---")
print('=' *80)
colunas = []
if not os.path.exists(HEADER_FILE_PATH):
    print(f"ERRO FATAL: Arquivo de cabe√ßalho '{HEADER_FILE_PATH}' n√£o encontrado. Verifique o caminho.")
    logging.error(f"Arquivo de cabe√ßalho '{HEADER_FILE_PATH}' n√£o encontrado.")
    raise FileNotFoundError(f"Arquivo de cabe√ßalho '{HEADER_FILE_PATH}' n√£o encontrado.")
else:
    header_df = pd.read_csv(HEADER_FILE_PATH, header=None, sep=";")
    colunas = header_df.iloc[0].tolist()
    print(f"Cabe√ßalhos carregados com sucesso. N√∫mero de colunas esperado: {len(colunas)}")
    logging.info(f"Cabe√ßalhos carregados: {colunas} | N√∫mero de colunas: {len(colunas)}")
print('-' *80)
print("ETAPA 2 Conclu√≠da: Cabe√ßalho carregado.\n")

# --- ETAPA 3: GERA√á√ÉO DE URLs ---
print('=' * 80)
print("--- Iniciando a ETAPA 3: Gera√ß√£o de URLs da Rede Sonda ---")
print('=' *80)
BASE_URL = "https://sonda.ccst.inpe.br/basedados/dados/sonda/eolicos/SCR"
ano_inicio = 2007
ano_fim = 2009
limites_meses = {2007: 12, 2008: 12, 2009: 9}
urls_dados = []

for ano in range(ano_inicio, ano_fim + 1):
    inicio_mes = 6 if ano == 2007 else 1
    fim_mes_ano = limites_meses.get(ano, 12)
    for mes in range(inicio_mes, fim_mes_ano + 1):
        aa = str(ano)[-2:]
        mm = f"{mes:02d}"
        ext = '7z' if ano == 2009 else 'zip'
        url = f"{BASE_URL}/{ano}/SCR{aa}{mm}WD.{ext}"
        urls_dados.append(url)

print(f"Total de {len(urls_dados)} URLs geradas para o per√≠odo de Jun/2007 a Set/2009.")
logging.info(f"Total de URLs geradas: {len(urls_dados)}")
print('-' * 80)
print("ETAPA 3 Conclu√≠da: URLs geradas.\n")

# --- ETAPA 4: DOWNLOAD, EXTRA√á√ÉO E SALVAMENTO CONDICIONAL (COM TRATAMENTO DE COLUNAS) ---
print('=' * 80)
print("--- Iniciando a ETAPA 4: Download, Extra√ß√£o e Salvamento Condicional ---")
print('=' *80)
# Adicionando a barra de progresso com tqdm
for url in tqdm(urls_dados, desc="Processando arquivos da Rede SONDA"):
    nome_arquivo_origem = url.split('/')[-1]
    csv_filename_destino = nome_arquivo_origem.replace('.zip', '.csv').replace('.7z', '.csv')
    csv_path_destino = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, csv_filename_destino)

    if os.path.exists(csv_path_destino):
        try:
            df_existente_check = pd.read_csv(csv_path_destino, sep=";", header=None, nrows=1, encoding="latin1")
            if df_existente_check.shape[1] != len(colunas):
                print(f"Arquivo '{csv_filename_destino}' ({df_existente_check.shape[1]} cols) com formato inv√°lido. Removendo para reprocessar.")
                logging.warning(f"Arquivo '{csv_filename_destino}' ({df_existente_check.shape[1]} cols) com formato de colunas diferente do esperado ({len(colunas)} cols). Removendo para reprocessamento.")
                os.remove(csv_path_destino)
            else:
                # print(f"Arquivo '{csv_filename_destino}' j√° existe e tem formato correto. Pulando.") # Removido para n√£o poluir o output com tqdm
                logging.info(f"Arquivo '{csv_filename_destino}' j√° existe e tem o formato correto ({len(colunas)} cols). Pulando download e processamento.")
                continue
        except pd.errors.EmptyDataError:
            print(f"Arquivo '{csv_filename_destino}' existe mas est√° vazio. Removendo para reprocessar.")
            logging.warning(f"Arquivo '{csv_filename_destino}' existe mas est√° vazio. Removendo para reprocessamento.")
            os.remove(csv_path_destino)
        except Exception as e:
            print(f"Erro ao verificar '{csv_filename_destino}': {e}. Removendo para reprocessar.")
            logging.error(f"Erro ao verificar arquivo existente '{csv_filename_destino}': {e}. Removendo para tentar reprocessamento.")
            try:
                os.remove(csv_path_destino)
            except OSError as oe:
                print(f"N√£o foi poss√≠vel remover '{csv_filename_destino}': {oe}")
                logging.error(f"N√£o foi poss√≠vel remover o arquivo '{csv_filename_destino}' para reprocessamento: {oe}")
    try:
        # print(f"Tentando baixar e processar: {url}") # Removido para n√£o poluir o output com tqdm
        logging.info(f"Baixando arquivo de: {url}")
        response = requests.get(url, timeout=60)
        response.raise_for_status()

        extensao_arquivo = nome_arquivo_origem.split('.')[-1].lower()
        df_processado = None

        if extensao_arquivo == 'zip':
            with zipfile.ZipFile(BytesIO(response.content)) as z:
                nomes_csv_no_zip = [f_name for f_name in z.namelist() if f_name.lower().endswith('.csv')]
                if not nomes_csv_no_zip:
                    print(f"AVISO: Nenhum .csv encontrado dentro do ZIP: {url}")
                    logging.warning(f"Nenhum arquivo .csv encontrado dentro do ZIP: {url}")
                    continue
                nome_csv_principal_zip = nomes_csv_no_zip[0]
                # print(f"  Processando '{nome_csv_principal_zip}' do ZIP '{nome_arquivo_origem}'...") # Removido
                logging.info(f"Processando '{nome_csv_principal_zip}' de dentro do ZIP '{nome_arquivo_origem}'")
                with z.open(nome_csv_principal_zip) as f:
                    df_processado = pd.read_csv(f, sep=";", header=None, encoding="latin1", low_memory=False)

        elif extensao_arquivo == '7z':
            temp_extract_dir = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, f"temp_extract_{nome_arquivo_origem}")
            os.makedirs(temp_extract_dir, exist_ok=True)
            logging.info(f"Criado diret√≥rio tempor√°rio para 7z: {temp_extract_dir}")
            try:
                with py7zr.SevenZipFile(BytesIO(response.content), mode='r') as archive:
                    nomes_arquivos_no_7z = archive.getnames()
                    nomes_csv_no_7z = [f_name for f_name in nomes_arquivos_no_7z if f_name.lower().endswith('.csv')]
                    if not nomes_csv_no_7z:
                        print(f"AVISO: Nenhum .csv encontrado dentro do 7z: {url}")
                        logging.warning(f"Nenhum arquivo .csv encontrado dentro do 7z: {url}")
                        continue
                    nome_csv_principal_7z = nomes_csv_no_7z[0]
                    # print(f"  Extraindo e processando '{nome_csv_principal_7z}' do 7z '{nome_arquivo_origem}'...") # Removido
                    logging.info(f"Extraindo '{nome_csv_principal_7z}' de dentro do 7z '{nome_arquivo_origem}' para {temp_extract_dir}")
                    archive.extract(path=temp_extract_dir, targets=[nome_csv_principal_7z])
                    caminho_csv_extraido_temp = os.path.join(temp_extract_dir, nome_csv_principal_7z)
                    df_processado = pd.read_csv(caminho_csv_extraido_temp, sep=";", header=None, encoding="latin1", low_memory=False)
            finally:
                if os.path.exists(temp_extract_dir):
                    shutil.rmtree(temp_extract_dir)
                    logging.info(f"Diret√≥rio tempor√°rio {temp_extract_dir} removido.")
        else:
            print(f"AVISO: Extens√£o de arquivo n√£o suportada '{extensao_arquivo}' para {url}")
            logging.warning(f"Extens√£o de arquivo n√£o suportada '{extensao_arquivo}' para {url}")
            continue

        if df_processado is not None:
            if "09" in csv_filename_destino and df_processado.shape[1] == 12:
                df_processado = df_processado.iloc[:, :-1]
                print(f"  Coluna extra removida de '{csv_filename_destino}' (original: {nome_arquivo_origem}).")
                logging.info(f"√öltima coluna removida do arquivo '{csv_filename_destino}' (original: {nome_arquivo_origem}), novo shape: {df_processado.shape}")

            if df_processado.shape[1] == len(colunas):
                df_processado.to_csv(csv_path_destino, index=False, header=False, sep=";", encoding="latin1")
                # print(f"  Arquivo salvo: '{csv_path_destino}' ({df_processado.shape[0]} linhas, {df_processado.shape[1]} colunas).") # Removido
                logging.info(f"Arquivo processado e salvo como: '{csv_path_destino}' com {df_processado.shape[0]} linhas e {df_processado.shape[1]} colunas.")
            else:
                print(f"ERRO: Ap√≥s processamento, '{csv_filename_destino}' (de {nome_arquivo_origem}) tem {df_processado.shape[1]} cols, esperado {len(colunas)}. N√ÉO SALVO.")
                logging.error(f"Ap√≥s processamento, arquivo '{csv_filename_destino}' (de {nome_arquivo_origem}) possui {df_processado.shape[1]} colunas, mas o esperado era {len(colunas)}. Arquivo n√£o salvo.")
        else:
            print(f"AVISO: Nenhum DataFrame foi processado para a URL {url}")
            logging.warning(f"Nenhum DataFrame foi processado para a URL {url}.")

    except requests.exceptions.RequestException as e:
        print(f"ERRO de requisi√ß√£o ao baixar {url}: {e}")
        logging.error(f"Erro de requisi√ß√£o ao baixar {url}: {e}")
    except zipfile.BadZipFile:
        print(f"ERRO: Arquivo ZIP de {url} inv√°lido/corrompido.")
        logging.error(f"Erro: Arquivo baixado de {url} (ZIP) n√£o √© v√°lido ou est√° corrompido.")
    except py7zr.exceptions.Bad7zFile:
        print(f"ERRO: Arquivo 7Z de {url} inv√°lido/corrompido.")
        logging.error(f"Erro: Arquivo baixado de {url} (7Z) n√£o √© v√°lido ou est√° corrompido.")
    except Exception as e:
        print(f"ERRO INESPERADO ao processar {url}: {str(e)}")
        logging.error(f"Erro inesperado ao processar {url}: {str(e)}")
print('-' * 80)
print("ETAPA 4 Conclu√≠da!!!\n")

# --- ETAPA 5: UNIFICA√á√ÉO DOS ARQUIVOS CSV INDIVIDUAIS (COM VERIFICA√á√ÉO) ---
print('=' * 80)
print("--- Iniciando a ETAPA 5: Unifica√ß√£o dos Arquivos CSV ---")
print('=' * 80)
arquivos_csv_individuais = sorted([f for f in os.listdir(INDIVIDUAL_CSV_SAVE_PATH) if f.lower().endswith('.csv') and os.path.isfile(os.path.join(INDIVIDUAL_CSV_SAVE_PATH, f))])
lista_dataframes_para_unir = []

if not arquivos_csv_individuais:
    print("AVISO: Nenhum arquivo CSV individual encontrado para unifica√ß√£o.")
    logging.warning("Nenhum arquivo CSV individual encontrado para unifica√ß√£o.")
else:
    print(f"Encontrados {len(arquivos_csv_individuais)} arquivos CSV para tentar unificar.")
    # Adicionando a barra de progresso com tqdm
    for nome_arquivo_csv in tqdm(arquivos_csv_individuais, desc="Unificando arquivos CSV"):
        caminho_completo_csv = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, nome_arquivo_csv)
        try:
            df_temp = pd.read_csv(caminho_completo_csv, sep=";", header=None, encoding="latin1", low_memory=False)
            if df_temp.shape[1] == len(colunas):
                lista_dataframes_para_unir.append(df_temp)
                # print(f"Lido arquivo '{nome_arquivo_csv}' para unifica√ß√£o ({df_temp.shape[0]} linhas).") # Removido
                logging.info(f"Arquivo '{nome_arquivo_csv}' ({df_temp.shape[0]} linhas, {df_temp.shape[1]} colunas) carregado para unifica√ß√£o.")
            else:
                print(f"ERRO: Arquivo '{nome_arquivo_csv}' tem {df_temp.shape[1]} cols, esperado {len(colunas)}. IGNORADO.")
                logging.error(f"Arquivo '{nome_arquivo_csv}' tem {df_temp.shape[1]} colunas, mas o esperado era {len(colunas)}. Ignorando este arquivo na unifica√ß√£o.")
        except pd.errors.EmptyDataError:
            print(f"AVISO: Arquivo '{nome_arquivo_csv}' est√° vazio. IGNORADO.")
            logging.warning(f"Arquivo '{nome_arquivo_csv}' est√° vazio. Ignorando.")
        except Exception as e:
            print(f"ERRO ao ler '{nome_arquivo_csv}': {str(e)}. IGNORADO.")
            logging.error(f"Erro ao ler o arquivo '{nome_arquivo_csv}' para unifica√ß√£o: {str(e)}")

dataset_unificado = pd.DataFrame()
if lista_dataframes_para_unir:
    dataset_unificado = pd.concat(lista_dataframes_para_unir, ignore_index=True)
    print(f"\nTotal de {len(lista_dataframes_para_unir)} arquivos CSV unificados com sucesso.")
    print(f"Dimens√µes do dataset unificado ANTES de adicionar cabe√ßalho: {dataset_unificado.shape}")
    logging.info(f"Dataset unificado criado com sucesso: {dataset_unificado.shape[0]} linhas x {dataset_unificado.shape[1]} colunas.")
else:
    print("\nNenhum DataFrame v√°lido foi carregado, o dataset unificado est√° vazio.")
    logging.warning("Nenhum DataFrame v√°lido foi carregado, o dataset unificado est√° vazio.")
print('-' * 80)
print("ETAPA 5 Conclu√≠da!!!\n")

# --- ETAPA 6: APLICAR CABE√áALHO, EXIBIR AMOSTRA E SALVAR ---
print('=' * 80)
print("--- Iniciando a ETAPA 6: Aplicar Cabe√ßalho, Exibir Amostra e Salvar Dataset Final ---")
print('=' * 80)
if not dataset_unificado.empty:
    if len(colunas) == dataset_unificado.shape[1]:
        try:
            dataset_unificado.columns = colunas
            print("Cabe√ßalho adicionado ao dataset unificado!")
            logging.info("Cabe√ßalho aplicado ao dataset unificado.")

            print("-" *80)
            print("Primeiras linhas do dataset com cabe√ßalho:")
            print("-" *80)
            display(dataset_unificado.head(1).round(2)) # Usando display para melhor formata√ß√£o

            num_duplicados = dataset_unificado.duplicated().sum()
            print("-" *80)
            print("\nCARACTER√çSTICAS DO DATAFRAME:")
            print(f" - Dimens√£o do dataframe: {dataset_unificado.shape}")
            print(f" - Dados duplicados      : {num_duplicados}")
            print("-" *80)
            print('\nDADOS NULOS:')
            print(dataset_unificado[['ws_50', 'wd_50']].isna().sum())
            print("-" *80)

            caminho_dataset_final = os.path.join(FINAL_DATASET_SAVE_PATH, FINAL_DATASET_NAME)
            if os.path.exists(caminho_dataset_final):
                print(f"\nDataset final '{FINAL_DATASET_NAME}' j√° existe em:\n'{FINAL_DATASET_SAVE_PATH}'.")
                print("Nenhum novo salvamento ser√° realizado.")
                logging.info(f"Dataset final '{FINAL_DATASET_NAME}' j√° existe em '{FINAL_DATASET_SAVE_PATH}'. Pulando salvamento.")
            else:
                dataset_unificado.to_csv(caminho_dataset_final, index=False, sep=";", encoding="latin1")
                print(f"\nDataset final salvo com sucesso em:\n'{caminho_dataset_final}'")
                logging.info(f"Dataset final salvo com sucesso em: '{caminho_dataset_final}'")
        except Exception as e:
            print(f"ERRO ao aplicar cabe√ßalho ou salvar o dataset final: {str(e)}")
            logging.error(f"Erro ao aplicar cabe√ßalho ou salvar o dataset final: {str(e)}")
            logging.error(f"Detalhes - Shape do dataset unificado: {dataset_unificado.shape}, N√∫mero de cabe√ßalhos: {len(colunas)}")
    else:
        print(f"ERRO FATAL: Falha ao aplicar cabe√ßalho. N√∫mero de colunas no cabe√ßalho ({len(colunas)}) n√£o corresponde ao dataset unificado ({dataset_unificado.shape[1]}).")
        logging.error(f"Falha ao aplicar cabe√ßalho: N√∫mero de colunas no cabe√ßalho ({len(colunas)}) n√£o corresponde ao dataset unificado ({dataset_unificado.shape[1]}).")
else:
    print("AVISO: Dataset unificado est√° vazio. Nenhuma a√ß√£o de salvamento ou aplica√ß√£o de cabe√ßalho realizada.")
    logging.warning("Dataset unificado est√° vazio. Nenhuma a√ß√£o de salvamento ou aplica√ß√£o de cabe√ßalho realizada.")

print('-' * 80)
print("ETAPA 6 Conclu√≠da. PROCESSO FINALIZADO!!!")
print("-" * 80)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado

# --- Fun√ß√£o Auxiliar (Repetida para Independ√™ncia) ---
def plot_boxplot_with_outliers_info_sonda(dataframe, column, title_prefix, xlabel_text, output_filename, color='orange'):
    """
    Gera um boxplot para uma coluna espec√≠fica do DataFrame da Rede SONDA,
    incluindo informa√ß√µes de outliers (IQR) e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o boxplot.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico (ex: "Boxplot -").
        xlabel_text (str): R√≥tulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        color (str, optional): Cor do boxplot. Padr√£o √© 'orange'.
    """
    print('=' * 100)
    print(f"--- Gerando Boxplot para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame SONDA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o boxplot.")
        logging.warning(f"Boxplot SONDA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"‚ö†Ô∏è AVISO: A s√©rie SONDA '{column}' est√° vazia ap√≥s remover NAs. N√£o foi poss√≠vel gerar o boxplot.")
        logging.warning(f"Boxplot SONDA n√£o gerado: S√©rie '{column}' vazia ap√≥s dropna.")
        print('-' * 100)
        return

    Q1 = data_series.quantile(0.25)
    Q3 = data_series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound_iqr = Q1 - 1.5 * IQR
    upper_bound_iqr = Q3 + 1.5 * IQR

    outliers_count = data_series[(data_series < lower_bound_iqr) | (data_series > upper_bound_iqr)].count()
    total_count = len(data_series)
    outliers_percentage = (outliers_count / total_count * 100) if total_count > 0 else 0

    plt.figure(figsize=(9, 3))
    sns.boxplot(x=data_series, color=color, linewidth=1.5, flierprops=dict(marker='o', markerfacecolor='red', markersize=5))

    plt.annotate(
        f'Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\nOutliers: {outliers_count} ({outliers_percentage:.2f}%)',
        xy=(0.95, 0.05),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='bottom'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Boxplot para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Aplica√ß√£o para dfSonda_jan2006_abr2007 ---
print('=' * 100)
print("--- Gerando gr√°ficos para dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)

# O DataFrame dfSonda_jan2006_abr2007 deve estar carregado e pr√©-processado nas c√©lulas anteriores.
# Exemplo de como garantir que ele esteja carregado, se esta c√©lula for executada isoladamente:
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/CieÃÇncia de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")
    # Garantir convers√£o de N/A e N/S para NaN (fundamental para as m√©tricas)
    for col in ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace({'N/A': np.nan, 'N/S': np.nan}).astype(float)
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfSonda_jan2006_abr2007: {str(e)}. Gr√°ficos SONDA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame() # Define como vazio para pular gr√°ficos

if not dfSonda_jan2006_abr2007.empty:
    # Gerar boxplot para ws_50 (SONDA)
    plot_boxplot_with_outliers_info_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50',
        'Boxplot - Velocidade do Vento 50m (SONDA)',
        'Velocidade (m/s)',
        './boxplot_SondaWS50.png',
        color='orange'
    )

    # Gerar boxplot para wd_50 (SONDA)
    plot_boxplot_with_outliers_info_sonda(
        dfSonda_jan2006_abr2007,
        'wd_50',
        'Boxplot - Dire√ß√£o do Vento 50m (SONDA)',
        'Dire√ß√£o (¬∞)',
        './boxplot_SondaWD50.png',
        color='lightcoral'
    )
else:
    print("Skipping SONDA boxplots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Boxplots para Rede SONDA Conclu√≠da. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado
from scipy.stats import norm # Para a curva de distribui√ß√£o normal

# Configurar logging (garanta que s√≥ √© configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Densidade ---
def plot_density_with_normal_sonda(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='skyblue', color_normal='red'):
    """
    Gera um gr√°fico de densidade para uma coluna espec√≠fica do DataFrame da Rede SONDA,
    incluindo uma curva de distribui√ß√£o normal te√≥rica e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o gr√°fico de densidade.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico (ex: "Distribui√ß√£o de Densidade -").
        xlabel_text (str): R√≥tulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        color_density (str, optional): Cor da densidade emp√≠rica. Padr√£o √© 'skyblue'.
        color_normal (str, optional): Cor da curva normal. Padr√£o √© 'red'.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Densidade para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame SONDA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade SONDA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"‚ö†Ô∏è AVISO: A s√©rie SONDA '{column}' est√° vazia ap√≥s remover NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade SONDA n√£o gerado: S√©rie '{column}' vazia ap√≥s dropna.")
        print('-' * 100)
        return

    # Calcular m√©dia e desvio padr√£o para a curva normal
    media = data_series.mean()
    desvio = data_series.std()

    plt.figure(figsize=(8, 5)) # Ajustei o tamanho para melhor visualiza√ß√£o
    sns.kdeplot(data_series, fill=True, label='Densidade Emp√≠rica', color=color_density, linewidth=2)

    # Gerar pontos para a curva normal te√≥rica
    x_normal = np.linspace(data_series.min(), data_series.max(), 200)
    y_normal = norm.pdf(x_normal, media, desvio)
    plt.plot(x_normal, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribui√ß√£o Normal Te√≥rica')

    # Adicionar linha vertical para a m√©dia
    plt.axvline(media, color='black', linestyle=':', linewidth=1.5, label=f'M√©dia: {media:.2f}')

    plt.annotate(
        f'M√©dia: {media:.2f}\nDesvio Padr√£o: {desvio:.2f}',
        xy=(0.95, 0.85), # Ajustado para n√£o sobrepor a legenda padr√£o
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade') # Adicionado r√≥tulo do eixo Y
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Densidade para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Carregamento do DataFrame dfSonda_jan2006_abr2007 ---
# Esta parte √© semelhante √† sua fun√ß√£o de boxplot para garantir que o dataframe esteja carregado
print('=' * 100)
print("--- Carregando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    # Garantir convers√£o de N/A e N/S para NaN e tipo float para colunas relevantes
    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            # Primeiro, substitui strings problem√°ticas por NaN
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan) # Adicionado -999 e -999.0 se forem usados para nulos
            # Tenta converter para float, os erros viram NaN
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')
    print("‚òë DataFrame carregado e colunas relevantes convertidas para num√©rico.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gr√°ficos.")

except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame() # Define como vazio para pular gr√°ficos
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfSonda_jan2006_abr2007: {str(e)}. Gr√°ficos SONDA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gr√°fico de densidade para ws_50 (SONDA)
    plot_density_with_normal_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50',
        'Distribui√ß√£o de Densidade (SONDA)',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_normal_SondaWS50_melhorado.png',
        color_density='deepskyblue', # Cor alterada para exemplo
        color_normal='crimson'      # Cor alterada para exemplo
    )
else:
    print("Skipping SONDA density plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Densidade para Rede SONDA Conclu√≠da. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado
from scipy.stats import weibull_min # Para a curva de distribui√ß√£o Weibull

# Configurar logging (garanta que s√≥ √© configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Densidade com Weibull ---
def plot_density_with_weibull_sonda(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='skyblue', color_weibull='green'):
    """
    Gera um gr√°fico de densidade para uma coluna espec√≠fica do DataFrame da Rede SONDA,
    incluindo uma curva de distribui√ß√£o Weibull te√≥rica e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o gr√°fico de densidade.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico (ex: "Distribui√ß√£o de Densidade -").
        xlabel_text (str): R√≥tulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        color_density (str, optional): Cor da densidade emp√≠rica. Padr√£o √© 'skyblue'.
        color_weibull (str, optional): Cor da curva Weibull. Padr√£o √© 'green'.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Densidade com Weibull para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame SONDA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Weibull SONDA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2: # weibull_min.fit precisa de pelo menos 2 pontos
        print(f"‚ö†Ô∏è AVISO: A s√©rie SONDA '{column}' est√° vazia ou tem poucos dados ap√≥s remover NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Weibull SONDA n√£o gerado: S√©rie '{column}' vazia ou com dados insuficientes ap√≥s dropna.")
        print('-' * 100)
        return

    # Ajustar a distribui√ß√£o de Weibull aos dados
    # floc=0 fixa o par√¢metro de localiza√ß√£o em 0, comum para velocidade do vento (n√£o pode ser negativa)
    try:
        shape, loc, scale = weibull_min.fit(data_series, floc=0)
    except Exception as e:
        print(f"‚ùå ERRO ao ajustar a distribui√ß√£o de Weibull para a coluna '{column}': {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}': {str(e)}")
        print('-' * 100)
        return

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data_series, fill=True, label='Densidade Emp√≠rica', color=color_density, linewidth=2)

    # Gerar pontos para a curva Weibull te√≥rica
    x_weibull = np.linspace(data_series.min(), data_series.max(), 200)
    y_weibull = weibull_min.pdf(x_weibull, shape, loc, scale)
    plt.plot(x_weibull, y_weibull, color_weibull, linestyle='--', linewidth=2, label='Distribui√ß√£o Weibull Te√≥rica')

    # Adicionar linha vertical para a m√©dia (opcional, mas pode ser √∫til para refer√™ncia)
    media = data_series.mean()
    plt.axvline(media, color='black', linestyle=':', linewidth=1.5, label=f'M√©dia Emp√≠rica: {media:.2f}')

    plt.annotate(
        f'Weibull Par√¢metros:\nShape (k): {shape:.2f}\nScale (Œª): {scale:.2f}\nLoc (Œ≥): {loc:.2f}', # loc ser√° 0 devido a floc=0
        xy=(0.95, 0.85),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Densidade Weibull para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Carregamento do DataFrame dfSonda_jan2006_abr2007 ---
# (Mesma l√≥gica de carregamento e pr√©-processamento da c√©lula anterior)
print('=' * 100)
print("--- Carregando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan)
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')
    print("‚òë DataFrame carregado e colunas relevantes convertidas para num√©rico.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gr√°ficos.")

except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfSonda_jan2006_abr2007: {str(e)}. Gr√°ficos SONDA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gr√°fico de densidade com Weibull para ws_50 (SONDA)
    plot_density_with_weibull_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50', # Foco no atributo ws_50
        'Distribui√ß√£o de Densidade (SONDA) com Ajuste Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_weibull_SondaWS50_melhorado.png',
        color_density='mediumseagreen', # Cor alterada para exemplo
        color_weibull='darkorange'      # Cor alterada para exemplo
    )
else:
    print("Skipping SONDA density plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Densidade com Weibull para Rede SONDA Conclu√≠da. ---")
print("-" * 100)
````

üìö **Distribui√ß√£o de Weibull:**

A distribui√ß√£o de Weibull √© uma distribui√ß√£o de probabilidade cont√≠nua que se destaca por sua versatilidade. Ela √© definida por dois ou tr√™s par√¢metros: o par√¢metro de forma **`(k)`**, o par√¢metro de escala **`(Œª ou c)`**, e, opcionalmente, um par√¢metro de localiza√ß√£o **`(Œ≥ ou Œ∏)`**.

- **Par√¢metro de Forma (k):** Determina a forma fundamental da fun√ß√£o de densidade de probabilidade. Valores diferentes de k podem fazer com que a distribui√ß√£o de Weibull se assemelhe a outras distribui√ß√µes. Por exemplo:
 - **`Se k < 1`**, a taxa de falha (ou, analogamente, a probabilidade de ocorr√™ncia de certas velocidades de vento) diminui com o tempo (ou velocidade).
 - **`Se k = 1`**, a distribui√ß√£o de Weibull se reduz a uma distribui√ß√£o exponencial, indicando uma taxa de falha constante.
 - **`Se k = 2`**, ela se torna uma distribui√ß√£o de Rayleigh, comum para descrever a distribui√ß√£o da velocidade do vento quando a dire√ß√£o do vento √© uniforme.
 - **`Se k est√° entre 3 e 4`**, a distribui√ß√£o de Weibull se aproxima da forma de uma distribui√ß√£o normal.

- **Par√¢metro de Escala (Œª):** Este par√¢metro "estica" ou "comprime" a distribui√ß√£o ao longo do eixo horizontal (neste caso, o eixo da velocidade do vento). Ele est√° diretamente relacionado √† velocidade m√©dia do vento; um valor maior de Œª geralmente implica velocidades m√©dias de vento mais altas.

- **Par√¢metro de Localiza√ß√£o (Œ≥):** Define o ponto de in√≠cio da distribui√ß√£o. Para a velocidade do vento, este par√¢metro √© frequentemente fixado em zero, pois a velocidade n√£o pode ser negativa.

 üí° A **`distribui√ß√£o de Weibull`** √© frequentemente empregada na an√°lise de recursos e√≥licos, pois se ajusta bem a uma ampla gama de dados de velocidade do vento observados em diferentes locais. Sua flexibilidade permite modelar distribui√ß√µes que podem ser sim√©tricas ou assim√©tricas, o que √© crucial para dados de vento que raramente seguem uma distribui√ß√£o normal perfeita. Conforme afirma a literatura especializada, "a fun√ß√£o de densidade de probabilidade de Weibull de dois par√¢metros tem sido amplamente utilizada na ind√∫stria de energia e√≥lica para descrever as distribui√ß√µes de velocidade do vento".

1. **Densidade Emp√≠rica (Curva Verde Preenchida):**

 - **Forma da Distribui√ß√£o:** A distribui√ß√£o emp√≠rica √© unimodal, com um pico claro e uma leve assimetria √† direita (cauda mais longa para velocidades de vento mais altas). Isso √© caracter√≠stico de dados de velocidade do vento, onde h√° um limite inferior **`(0 m/s)`** e a possibilidade de ocorr√™ncia de velocidades mais elevadas, embora menos frequentes.

 - **Pico de Densidade:** O ponto mais alto da curva de densidade emp√≠rica, indicando a faixa de velocidade do vento mais comum, localiza-se em torno de **`4.5 m/s`**.

 - **Dispers√£o:** A maior parte das observa√ß√µes de velocidade do vento concentra-se entre aproximadamente 2 m/s e 8 m/s.

2. **Distribui√ß√£o Weibull Te√≥rica (Linha Laranja Tracejada):**

 - **Ajuste aos Dados:** A curva de Weibull te√≥rica, cujos par√¢metros estimados s√£o **`Shape (k) = 2.53, Scale (Œª) = 5.94 m/s, e Location (Œ≥) = 0.00 m/s (fixado)`**, demonstra um bom ajuste √† distribui√ß√£o emp√≠rica. Ela acompanha bem o formato geral da curva dos dados, capturando o pico e a inclina√ß√£o das caudas, especialmente a cauda direita, de forma mais eficaz do que uma distribui√ß√£o normal simples faria para este tipo de dado.

 - **Par√¢metros de Weibull Anotados:**
    - **`Shape (k) = 2.53:`** Este par√¢metro (adimensional) indica a forma da distribui√ß√£o. Um valor de k > 2 (como o obtido) sugere uma distribui√ß√£o que se assemelha a uma curva de sino, embora possa ser assim√©trica. Se k=2, a distribui√ß√£o de Weibull torna-se uma distribui√ß√£o de **`Rayleigh`**, frequentemente usada para dados de vento.
    - **`Scale (Œª) = 5.94 m/s:`** O par√¢metro de escala est√° relacionado com a velocidade m√©dia do vento e influencia a dispers√£o da distribui√ß√£o ao longo do eixo da velocidade.
    - **`Location (Œ≥) = 0.00 m/s:`** O par√¢metro de localiza√ß√£o foi fixado em zero, o que √© fisicamente apropriado para a velocidade do vento, pois n√£o pode assumir valores negativos.

3. **M√©dia Emp√≠rica (Linha Preta Pontilhada):**

 - **Valor e Posi√ß√£o:** A m√©dia emp√≠rica da velocidade do vento √© de **`5.28 m/s`**. Como observado no gr√°fico anterior, ela se localiza √† direita do pico da densidade emp√≠rica (e tamb√©m do pico da curva de Weibull ajustada), o que √© consistente com a assimetria √† direita da distribui√ß√£o.

#### üìà **Gr√°fico de Densidade com curva Normal e Weibull**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado
from scipy.stats import norm, weibull_min # Para as curvas de distribui√ß√£o

# Configurar logging (garanta que s√≥ √© configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Densidade Combinado ---
def plot_density_combined_sonda(dataframe, column, title_prefix, xlabel_text, output_filename,
                                color_density='skyblue', color_normal='red', color_weibull='green'):
    """
    Gera um gr√°fico de densidade para uma coluna espec√≠fica do DataFrame da Rede SONDA,
    incluindo curvas te√≥ricas Normal e Weibull, e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o gr√°fico de densidade.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico.
        xlabel_text (str): R√≥tulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        color_density (str, optional): Cor da densidade emp√≠rica.
        color_normal (str, optional): Cor da curva normal.
        color_weibull (str, optional): Cor da curva Weibull.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Densidade Combinado para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame SONDA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Combinado SONDA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return None, None # Retorna None para os par√¢metros

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2:
        print(f"‚ö†Ô∏è AVISO: A s√©rie SONDA '{column}' est√° vazia ou tem poucos dados ap√≥s remover NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Combinado SONDA n√£o gerado: S√©rie '{column}' vazia ou com dados insuficientes.")
        print('-' * 100)
        return None, None # Retorna None para os par√¢metros

    # Par√¢metros para a curva Normal
    media_normal = data_series.mean()
    desvio_normal = data_series.std()
    params_normal = {'media': media_normal, 'desvio': desvio_normal}

    # Ajustar a distribui√ß√£o de Weibull
    try:
        shape_weibull, loc_weibull, scale_weibull = weibull_min.fit(data_series, floc=0)
        params_weibull = {'shape': shape_weibull, 'loc': loc_weibull, 'scale': scale_weibull}
    except Exception as e:
        print(f"‚ùå ERRO ao ajustar a distribui√ß√£o de Weibull para a coluna '{column}': {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}': {str(e)}")
        params_weibull = None # Indica que o ajuste Weibull falhou
        print('-' * 100)
        # Continuamos para plotar pelo menos a Normal, se poss√≠vel

    plt.figure(figsize=(9, 6)) # Ajustado para melhor visualiza√ß√£o com mais elementos
    sns.kdeplot(data_series, fill=True, label='Densidade Emp√≠rica', color=color_density, linewidth=2, alpha=0.6)

    # Gerar pontos para as curvas te√≥ricas
    x_vals = np.linspace(data_series.min(), data_series.max(), 200)

    # Curva Normal
    y_normal = norm.pdf(x_vals, media_normal, desvio_normal)
    plt.plot(x_vals, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribui√ß√£o Normal Te√≥rica')

    # Curva Weibull (se o ajuste foi bem-sucedido)
    if params_weibull:
        y_weibull = weibull_min.pdf(x_vals, shape_weibull, loc_weibull, scale_weibull)
        plt.plot(x_vals, y_weibull, color_weibull, linestyle=':', linewidth=2, label='Distribui√ß√£o Weibull Te√≥rica')

    # Linha vertical para a m√©dia emp√≠rica
    plt.axvline(media_normal, color='black', linestyle='-.', linewidth=1, label=f'M√©dia Emp√≠rica: {media_normal:.2f}')

    # Anota√ß√µes
    annotation_text = f'Normal:\nM√©dia: {media_normal:.2f}\nDesvio Padr√£o: {desvio_normal:.2f}\n'
    if params_weibull:
        annotation_text += f'\nWeibull:\nShape (k): {shape_weibull:.2f}\nScale (Œª): {scale_weibull:.2f}\nLoc (Œ≥): {loc_weibull:.2f}'
    else:
        annotation_text += "\nWeibull: Ajuste falhou"

    plt.annotate(
        annotation_text,
        xy=(0.95, 0.95),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.9),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend(loc='upper left', bbox_to_anchor=(0.01, 0.95)) # Ajuste da posi√ß√£o da legenda
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Densidade Combinado para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)
    return params_normal, params_weibull


# --- Carregamento do DataFrame dfSonda_jan2006_abr2007 ---
# (Mesma l√≥gica de carregamento e pr√©-processamento da c√©lula anterior)
print('=' * 100)
print("--- Carregando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan)
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')
    print("‚òë DataFrame carregado e colunas relevantes convertidas para num√©rico.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gr√°ficos.")

except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfSonda_jan2006_abr2007: {str(e)}. Gr√°ficos SONDA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gr√°fico de densidade combinado para ws_50 (SONDA)
    params_normal_ws50, params_weibull_ws50 = plot_density_combined_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50', # Foco no atributo ws_50
        'Densidade - SONDAWS50: Emp√≠rica vs Normal vs Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_Normal_Weibull_SondaWS50_melhorado.png'
    )

    # Exibir par√¢metros estimados se foram calculados
    if params_normal_ws50:
        print(f"\nüìä Par√¢metros Estimados para Distribui√ß√£o Normal (ws_50):")
        print(f"   - M√©dia: {params_normal_ws50['media']:.2f}")
        print(f"   - Desvio Padr√£o: {params_normal_ws50['desvio']:.2f}")
    if params_weibull_ws50:
        print(f"\nüìà Par√¢metros Estimados para Distribui√ß√£o Weibull (ws_50):")
        print(f"   - Shape (k): {params_weibull_ws50['shape']:.2f}")
        print(f"   - Scale (Œª): {params_weibull_ws50['scale']:.2f}")
        print(f"   - Loc (Œ≥): {params_weibull_ws50['loc']:.2f}") # Ser√° 0 se floc=0 foi usado
else:
    print("Skipping SONDA density plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Densidade Combinados para Rede SONDA Conclu√≠da. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado

# Configurar logging (garanta que s√≥ √© configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Dispers√£o Temporal ---
def plot_temporal_dispersion_sonda(dataframe, time_column, data_column, title_prefix, xlabel_text, ylabel_text, output_filename,
                                   raw_data_color='skyblue', trend_color='red', raw_data_label_suffix='(Dados Hor√°rios)'):
    """
    Gera um gr√°fico de dispers√£o temporal para uma coluna espec√≠fica do DataFrame da Rede SONDA,
    incluindo uma linha de tend√™ncia LOESS e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        time_column (str): Nome da coluna de data/hora.
        data_column (str): Nome da coluna de dados para plotar.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico.
        xlabel_text (str): R√≥tulo do eixo X.
        ylabel_text (str): R√≥tulo do eixo Y.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        raw_data_color (str, optional): Cor dos dados brutos.
        trend_color (str, optional): Cor da linha de tend√™ncia.
        raw_data_label_suffix (str, optional): Sufixo para a legenda dos dados brutos.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Dispers√£o Temporal para SONDA: {data_column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or time_column not in dataframe.columns or data_column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame SONDA est√° vazio ou as colunas '{time_column}' ou '{data_column}' n√£o existem. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Dispers√£o SONDA n√£o gerado: DataFrame vazio ou colunas ausentes.")
        print('-' * 100)
        return

    # Certificar que a coluna de tempo √© do tipo datetime
    try:
        df_plot = dataframe.copy() # Trabalhar com uma c√≥pia para evitar SettingWithCopyWarning
        df_plot[time_column] = pd.to_datetime(df_plot[time_column])
    except Exception as e:
        print(f"‚ùå ERRO ao converter a coluna de tempo '{time_column}' para datetime: {str(e)}")
        logging.error(f"Erro ao converter '{time_column}' para datetime: {str(e)}")
        print('-' * 100)
        return

    data_series = df_plot[data_column].dropna()
    time_series = df_plot.loc[data_series.index, time_column] # Alinhar tempo com dados n√£o nulos

    if data_series.empty:
        print(f"‚ö†Ô∏è AVISO: A s√©rie SONDA '{data_column}' est√° vazia ap√≥s remover NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Dispers√£o SONDA n√£o gerado: S√©rie '{data_column}' vazia ap√≥s dropna.")
        print('-' * 100)
        return

    plt.figure(figsize=(12, 5)) # Ajustado para melhor visualiza√ß√£o temporal

    # Plotar os dados brutos (pontos menores e com transpar√™ncia devido √† densidade)
    # Usar plt.plot para dados de alta frequ√™ncia pode ser mais perform√°tico e visualmente limpo
    plt.plot(time_series, data_series, label=f'{data_column} {raw_data_label_suffix}', color=raw_data_color, alpha=0.6, linestyle='-', marker=None, linewidth=0.8)

    # Linha de tend√™ncia LOESS
    # regplot precisa de valores num√©ricos para x se n√£o interpretar datetime diretamente
    # Convertendo para n√∫meros de data matplotlib se necess√°rio, embora seaborn mais recente lide melhor com datetime
    time_numeric = mdates.date2num(time_series)
    sns.regplot(
        x=time_numeric,
        y=data_series,
        scatter=False, # J√° plotamos os dados brutos
        lowess=True,   # Usa LOESS para uma tend√™ncia mais suave e n√£o linear
        color=trend_color,
        label='Tend√™ncia (LOESS)',
        line_kws={'linewidth': 2}
    )

    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=12))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45, ha='right')

    plt.title(f'{title_prefix} - {data_column} (01/2006 a 04/2007)')
    plt.xlabel(xlabel_text)
    plt.ylabel(ylabel_text)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Dispers√£o Temporal para SONDA '{data_column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Carregamento e Pr√©-processamento do DataFrame dfSonda_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando e Pr√©-processando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    # Converter 'datetm' para datetime, crucial para plots temporais
    dfSonda_jan2006_abr2007['datetm'] = pd.to_datetime(dfSonda_jan2006_abr2007['datetm'], errors='coerce')

    # Garantir convers√£o de N/A e N/S para NaN e tipo float para colunas de medi√ß√£o
    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan)
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')

    # Remover linhas onde 'datetm' ou 'ws_50' s√£o NaT/NaN ap√≥s convers√£o, para evitar problemas no plot
    dfSonda_jan2006_abr2007.dropna(subset=['datetm', 'ws_50'], inplace=True)

    print("‚òë DataFrame carregado e colunas relevantes convertidas/limpas.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gr√°ficos.")

except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' n√£o encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfSonda_jan2006_abr2007: {str(e)}. Gr√°ficos SONDA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gr√°fico de dispers√£o temporal para ws_50 (SONDA)
    plot_temporal_dispersion_sonda(
        dfSonda_jan2006_abr2007,
        time_column='datetm',       # Coluna de tempo
        data_column='ws_50',        # Coluna de dados
        title_prefix='Dispers√£o Temporal (SONDA)',
        xlabel_text='Tempo',
        ylabel_text='Velocidade do Vento (m/s) a 50m',
        output_filename='./dispersao_temporal_SondaWS50_melhorado.png',
        raw_data_label_suffix='(Dados 10 min)' # Ajustado para refletir a frequ√™ncia original
    )
else:
    print("Skipping SONDA dispersion plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Dispers√£o Temporal para Rede SONDA Conclu√≠da. ---")
print("-" * 100)
````

#### <H3><b> ‚úÖ An√°lise do Gr√°fico: Dispers√£o Temporal (SONDA) - ws_50 (01/2006 a 04/2007)</b>

- Este gr√°fico ilustra a varia√ß√£o da velocidade do vento a 50 metros **`(ws_50)`** ao longo do tempo, com dados coletados pela Rede SONDA **`(a cada 10 minutos)`** no per√≠odo de janeiro de 2006 a abril de 2007. Uma linha de tend√™ncia **`LOESS (Locally Estimated Scatterplot Smoothing)`** foi sobreposta para suavizar as flutua√ß√µes e destacar padr√µes de longo prazo.


<H3><b>Observa√ß√µes Detalhadas do Gr√°fico:</b>

1. **Dados Brutos (ws_50 - Dados 10 min - √Årea Azul Clara):**

- **`Variabilidade de Curto Prazo:`** A √°rea azul clara representa a densidade dos dados brutos, medidos a cada 10 minutos. Observa-se uma consider√°vel variabilidade e "ru√≠do" em curtos per√≠odos, o que √© t√≠pico para medi√ß√µes de velocidade do vento devido a rajadas, calmarias e outros fen√¥menos meteorol√≥gicos de curta dura√ß√£o.
- **`Amplitude das Medi√ß√µes:`** As velocidades do vento variam significativamente, com picos ocasionais que ultrapassam os 10 m/s e per√≠odos de ventos mais fracos, pr√≥ximos a `0-2 m/s`. A maior densidade de pontos parece se concentrar na faixa de `2 m/s a 8 m/s`.

2. **Linha de Tend√™ncia (LOESS - Linha Vermelha):**

- **`Padr√£o Sazonal/Temporal:`** A linha de tend√™ncia LOESS revela um padr√£o mais suave e claro ao longo do per√≠odo analisado:
 - `In√≠cio de 2006 (Jan-Fev):` A tend√™ncia come√ßa em torno de 5 m/s.
 - `Meados de 2006 (Mar-Jul):` H√° uma leve queda na tend√™ncia, atingindo um vale pr√≥ximo a 4.5 m/s por volta de julho de 2006.

 - `In√≠cio de 2007 (Jan-Mai):` Ap√≥s o pico no final de 2006, a tend√™ncia indica uma diminui√ß√£o gradual da velocidade m√©dia do vento, caindo para valores pr√≥ximos a 4.5 m/s ao final do per√≠odo observado (abril/maio de 2007).

- **`Suaviza√ß√£o:`** A curva LOESS efetivamente filtra as flutua√ß√µes de alta frequ√™ncia, permitindo visualizar a tend√™ncia geral da velocidade do vento, que parece apresentar um ciclo anual com velocidades mais altas no segundo semestre do ano.

<H3><b>Conclus√µes e Insights:</b>

- **`Comportamento Temporal do Vento:`** O gr√°fico demonstra que a velocidade do vento na localidade n√£o √© constante, exibindo varia√ß√µes di√°rias (`impl√≠citas na densidade dos dados brutos`) e tend√™ncias de mais longo prazo (`sazonais ou interanuais`), como o aumento no segundo semestre de 2006.
- **`Utilidade da Linha de Tend√™ncia LOESS:`** A linha LOESS √© particularmente √∫til para dados de alta frequ√™ncia como estes, pois ajuda a identificar padr√µes que poderiam ser obscurecidos pela volatilidade dos dados brutos. Ela sugere um comportamento c√≠clico ou sazonal que pode ser explorado em modelos de previs√£o.
- **`Qualidade dos Dados para Modelagem:`** A visualiza√ß√£o dos dados brutos, apesar de sua variabilidade, n√£o mostra grandes lacunas ou anomalias abruptas que n√£o sejam capturadas pela linha de tend√™ncia como parte da variabilidade natural do vento.
- **`Refer√™ncia para Per√≠odos Faltantes:`** A tend√™ncia observada pode ser √∫til como uma primeira aproxima√ß√£o ou para validar imputa√ß√µes para per√≠odos com dados faltantes, como o m√™s de maio de 2007 que √© o foco do projeto.

---
üìö **O que √© a Curva LOESS?**

>LOESS √© uma t√©cnica de regress√£o n√£o param√©trica robusta que visa criar uma curva suave atrav√©s de um gr√°fico de dispers√£o de pontos. Diferentemente de uma regress√£o linear simples, que ajusta uma √∫nica reta a todos os dados, o LOESS ajusta m√∫ltiplos modelos de regress√£o simples a subconjuntos localizados dos dados. Isso o torna particularmente eficaz para revelar tend√™ncias e ciclos em dados que podem n√£o seguir um padr√£o linear global ou uma fun√ß√£o matem√°tica espec√≠fica.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado

# --- Fun√ß√£o Auxiliar (Repetida para Independ√™ncia) ---
def plot_boxplot_with_outliers_info_nasa(dataframe, column, title_prefix, xlabel_text, output_filename, color='skyblue'):
    """
    Gera um boxplot para uma coluna espec√≠fica do DataFrame da NASA POWER,
    incluindo informa√ß√µes de outliers (IQR) e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da NASA.
        column (str): O nome da coluna para gerar o boxplot.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico (ex: "Boxplot -").
        xlabel_text (str): R√≥tulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        color (str, optional): Cor do boxplot. Padr√£o √© 'skyblue'.
    """
    print('=' * 100)
    print(f"--- Gerando Boxplot para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame NASA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o boxplot.")
        logging.warning(f"Boxplot NASA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"‚ö†Ô∏è AVISO: A s√©rie NASA '{column}' est√° vazia ap√≥s remover NAs. N√£o foi poss√≠vel gerar o boxplot.")
        logging.warning(f"Boxplot NASA n√£o gerado: S√©rie '{column}' vazia ap√≥s dropna.")
        print('-' * 100)
        return

    Q1 = data_series.quantile(0.25)
    Q3 = data_series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound_iqr = Q1 - 1.5 * IQR
    upper_bound_iqr = Q3 + 1.5 * IQR

    outliers_count = data_series[(data_series < lower_bound_iqr) | (data_series > upper_bound_iqr)].count()
    total_count = len(data_series)
    outliers_percentage = (outliers_count / total_count * 100) if total_count > 0 else 0

    plt.figure(figsize=(9, 3))
    sns.boxplot(x=data_series, color=color, linewidth=1.5, flierprops=dict(marker='o', markerfacecolor='red', markersize=5))

    plt.annotate(
        f'Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\nOutliers: {outliers_count} ({outliers_percentage:.2f}%)',
        xy=(0.95, 0.05),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='bottom'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Boxplot para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Aplica√ß√£o para dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Gerando gr√°ficos para dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)

# O DataFrame dfNASA_jan2006_abr2007 deve estar carregado nas c√©lulas anteriores.
# Exemplo de como garantir que ele esteja carregado, se esta c√©lula for executada isoladamente:
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/CieÃÇncia de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
except Exception as e:
    print(f"‚ùå ERRO ao carregar dfNASA_jan2006_abr2007: {str(e)}. Gr√°ficos NASA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos NASA: {str(e)}")
    dfNASA_jan2006_abr2007 = pd.DataFrame() # Define como vazio para pular gr√°ficos

if not dfNASA_jan2006_abr2007.empty:
    # Gerar boxplot para WS50M (NASA)
    plot_boxplot_with_outliers_info_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M',
        'Boxplot - Velocidade do Vento 50m (NASA)',
        'Velocidade (m/s)',
        './boxplot_NASAWS50.png',
        color='skyblue'
    )

    # Gerar boxplot para WD50M (NASA)
    plot_boxplot_with_outliers_info_nasa(
        dfNASA_jan2006_abr2007,
        'WD50M',
        'Boxplot - Dire√ß√£o do Vento 50m (NASA)',
        'Dire√ß√£o (¬∞)',
        './boxplot_NASAWD50.png',
        color='lightgreen'
    )
else:
    print("Skipping NASA boxplots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Boxplots para NASA POWER Conclu√≠da. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas est√° importado
import logging # Garanta que logging est√° importado
from IPython.display import display # Garanta que display est√° importado
from scipy.stats import norm # Para a curva de distribui√ß√£o normal

# Configurar logging (garanta que s√≥ √© configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Densidade (NASA) ---
def plot_density_with_normal_nasa(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='lightcoral', color_normal='darkred'):
    """
    Gera um gr√°fico de densidade para uma coluna espec√≠fica do DataFrame da NASA,
    incluindo uma curva de distribui√ß√£o normal te√≥rica e salva o gr√°fico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da NASA.
        column (str): O nome da coluna para gerar o gr√°fico de densidade.
        title_prefix (str): Prefixo para o t√≠tulo do gr√°fico.
        xlabel_text (str): R√≥tulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gr√°fico.
        color_density (str, optional): Cor da densidade emp√≠rica.
        color_normal (str, optional): Cor da curva normal.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Densidade para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame NASA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade NASA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"‚ö†Ô∏è AVISO: A s√©rie NASA '{column}' est√° vazia ap√≥s remover NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade NASA n√£o gerado: S√©rie '{column}' vazia ap√≥s dropna.")
        print('-' * 100)
        return

    media = data_series.mean()
    desvio = data_series.std()

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data_series, fill=True, label='Densidade Emp√≠rica', color=color_density, linewidth=2, alpha=0.7)

    x_normal = np.linspace(data_series.min(), data_series.max(), 200)
    y_normal = norm.pdf(x_normal, media, desvio)
    plt.plot(x_normal, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribui√ß√£o Normal Te√≥rica')

    plt.axvline(media, color='black', linestyle=':', linewidth=1.5, label=f'M√©dia: {media:.2f}')

    plt.annotate(
        f'M√©dia: {media:.2f}\nDesvio Padr√£o: {desvio:.2f}',
        xy=(0.95, 0.85),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Densidade para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)

    # Certificar que a coluna WS50M √© num√©rica (j√° deve ser, mas √© uma boa pr√°tica)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns:
        dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')

    print("‚òë DataFrame NASA carregado e coluna WS50M verificada/convertida para num√©rico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gr√°ficos.")

except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' n√£o encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' n√£o encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfNASA_jan2006_abr2007: {str(e)}. Gr√°ficos NASA n√£o gerados.")
    logging.error(f"Erro no setup para gr√°ficos NASA: {str(e)}")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    plot_density_with_normal_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M', # Coluna da NASA para velocidade do vento a 50m
        'Distribui√ß√£o de Densidade (NASA)',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_normal_NASAWS50_melhorado.png'
    )
else:
    print("Skipping NASA density plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Densidade com Normal para NASA POWER Conclu√≠da. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd
import logging
from IPython.display import display
from scipy.stats import weibull_min # Para a curva de distribui√ß√£o Weibull

# Configurar logging
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Densidade com Weibull (NASA) ---
def plot_density_with_weibull_nasa(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='lightgreen', color_weibull='darkgreen'):
    """
    Gera um gr√°fico de densidade para uma coluna espec√≠fica do DataFrame da NASA,
    incluindo uma curva de distribui√ß√£o Weibull te√≥rica e salva o gr√°fico.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Densidade com Weibull para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame NASA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Weibull NASA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2:
        print(f"‚ö†Ô∏è AVISO: A s√©rie NASA '{column}' est√° vazia ou tem poucos dados ap√≥s remover NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Weibull NASA n√£o gerado: S√©rie '{column}' vazia ou com dados insuficientes.")
        print('-' * 100)
        return

    try:
        shape, loc, scale = weibull_min.fit(data_series, floc=0)
    except Exception as e:
        print(f"‚ùå ERRO ao ajustar a distribui√ß√£o de Weibull para a coluna '{column}': {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}' (NASA): {str(e)}")
        print('-' * 100)
        return

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data_series, fill=True, label='Densidade Emp√≠rica', color=color_density, linewidth=2, alpha=0.7)

    x_weibull = np.linspace(data_series.min(), data_series.max(), 200)
    y_weibull = weibull_min.pdf(x_weibull, shape, loc, scale)
    plt.plot(x_weibull, y_weibull, color_weibull, linestyle='--', linewidth=2, label='Distribui√ß√£o Weibull Te√≥rica')

    media_empirica = data_series.mean()
    plt.axvline(media_empirica, color='black', linestyle=':', linewidth=1.5, label=f'M√©dia Emp√≠rica: {media_empirica:.2f}')

    plt.annotate(
        f'Weibull Par√¢metros:\nShape (k): {shape:.2f}\nScale (Œª): {scale:.2f}\nLoc (Œ≥): {loc:.2f}',
        xy=(0.95, 0.85),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Densidade Weibull para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns:
        dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')
    print("‚òë DataFrame NASA carregado e coluna WS50M verificada/convertida para num√©rico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gr√°ficos.")
except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' n√£o encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfNASA_jan2006_abr2007: {str(e)}. Gr√°ficos NASA n√£o gerados.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    plot_density_with_weibull_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M',
        'Distribui√ß√£o de Densidade (NASA) com Ajuste Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_weibull_NASAWS50_melhorado.png'
    )
else:
    print("Skipping NASA density plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Densidade com Weibull para NASA POWER Conclu√≠da. ---")
print("-" * 100)
````

#### üìà **Gr√°fico de Densidade com curva Normal e Weibull**
---

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd
import logging
from IPython.display import display
from scipy.stats import norm, weibull_min

# Configurar logging
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Densidade Combinado (NASA) ---
def plot_density_combined_nasa(dataframe, column, title_prefix, xlabel_text, output_filename,
                               color_density='palegreen', color_normal='darkred', color_weibull='darkorange'):
    """
    Gera um gr√°fico de densidade para uma coluna espec√≠fica do DataFrame da NASA,
    incluindo curvas te√≥ricas Normal e Weibull, e salva o gr√°fico.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Densidade Combinado para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"‚ö†Ô∏è AVISO: DataFrame NASA est√° vazio ou a coluna '{column}' n√£o existe. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Combinado NASA n√£o gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return None, None

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2:
        print(f"‚ö†Ô∏è AVISO: A s√©rie NASA '{column}' est√° vazia ou tem poucos dados. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Densidade Combinado NASA n√£o gerado: S√©rie '{column}' vazia ou com dados insuficientes.")
        print('-' * 100)
        return None, None

    media_normal = data_series.mean()
    desvio_normal = data_series.std()
    params_normal = {'media': media_normal, 'desvio': desvio_normal}

    try:
        shape_weibull, loc_weibull, scale_weibull = weibull_min.fit(data_series, floc=0)
        params_weibull = {'shape': shape_weibull, 'loc': loc_weibull, 'scale': scale_weibull}
    except Exception as e:
        print(f"‚ùå ERRO ao ajustar a distribui√ß√£o de Weibull para '{column}' (NASA): {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}' (NASA): {str(e)}")
        params_weibull = None
        print('-' * 100)

    plt.figure(figsize=(9, 6))
    sns.kdeplot(data_series, fill=True, label='Densidade Emp√≠rica', color=color_density, linewidth=2, alpha=0.6)

    x_vals = np.linspace(data_series.min(), data_series.max(), 200)

    y_normal = norm.pdf(x_vals, media_normal, desvio_normal)
    plt.plot(x_vals, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribui√ß√£o Normal Te√≥rica')

    if params_weibull:
        y_weibull = weibull_min.pdf(x_vals, shape_weibull, loc_weibull, scale_weibull)
        plt.plot(x_vals, y_weibull, color_weibull, linestyle=':', linewidth=2, label='Distribui√ß√£o Weibull Te√≥rica')

    plt.axvline(media_normal, color='black', linestyle='-.', linewidth=1, label=f'M√©dia Emp√≠rica: {media_normal:.2f}')

    annotation_text = f'Normal:\nM√©dia: {media_normal:.2f}\nDesvio Padr√£o: {desvio_normal:.2f}\n'
    if params_weibull:
        annotation_text += f'\nWeibull:\nShape (k): {shape_weibull:.2f}\nScale (Œª): {scale_weibull:.2f}\nLoc (Œ≥): {loc_weibull:.2f}'
    else:
        annotation_text += "\nWeibull: Ajuste falhou"

    plt.annotate(
        annotation_text,
        xy=(0.95, 0.95),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.9),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend(loc='upper left', bbox_to_anchor=(0.01, 0.95))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Densidade Combinado para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)
    return params_normal, params_weibull

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns:
        dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')
    print("‚òë DataFrame NASA carregado e coluna WS50M verificada/convertida para num√©rico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gr√°ficos.")
except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' n√£o encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfNASA_jan2006_abr2007: {str(e)}. Gr√°ficos NASA n√£o gerados.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    params_normal_nasa_ws50, params_weibull_nasa_ws50 = plot_density_combined_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M',
        'Densidade - NASA WS50M: Emp√≠rica vs Normal vs Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_Normal_Weibull_NASAWS50_melhorado.png'
    )

    if params_normal_nasa_ws50:
        print(f"\nüìä Par√¢metros Estimados para Distribui√ß√£o Normal (WS50M - NASA):")
        print(f"   - M√©dia: {params_normal_nasa_ws50['media']:.2f}")
        print(f"   - Desvio Padr√£o: {params_normal_nasa_ws50['desvio']:.2f}")
    if params_weibull_nasa_ws50:
        print(f"\nüìà Par√¢metros Estimados para Distribui√ß√£o Weibull (WS50M - NASA):")
        print(f"   - Shape (k): {params_weibull_nasa_ws50['shape']:.2f}")
        print(f"   - Scale (Œª): {params_weibull_nasa_ws50['scale']:.2f}")
        print(f"   - Loc (Œ≥): {params_weibull_nasa_ws50['loc']:.2f}")
else:
    print("Skipping NASA combined density plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Densidade Combinados para NASA POWER Conclu√≠da. ---")
print("-" * 100)
````

#### üìà **Gr√°fico de Dispers√£o Temporal - Velocidade do Vento a 50m**

````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import os
import pandas as pd
import logging
from IPython.display import display

# Configurar logging
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Fun√ß√£o Auxiliar Melhorada para Gr√°fico de Dispers√£o Temporal (NASA) ---
def plot_temporal_dispersion_nasa(dataframe, year_col, mo_col, dy_col, hr_col, data_column,
                                  title_prefix, xlabel_text, ylabel_text, output_filename,
                                  raw_data_color='dodgerblue', trend_color='darkorange', raw_data_label_suffix='(Dados Hor√°rios)'):
    """
    Gera um gr√°fico de dispers√£o temporal para uma coluna espec√≠fica do DataFrame da NASA,
    incluindo uma linha de tend√™ncia LOESS e salva o gr√°fico.
    """
    print('=' * 100)
    print(f"--- Gerando Gr√°fico de Dispers√£o Temporal para NASA: {data_column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or not all(col in dataframe.columns for col in [year_col, mo_col, dy_col, hr_col, data_column]):
        print(f"‚ö†Ô∏è AVISO: DataFrame NASA est√° vazio ou colunas de tempo/dados n√£o existem. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Dispers√£o NASA n√£o gerado: DataFrame vazio ou colunas ausentes.")
        print('-' * 100)
        return

    df_plot = dataframe.copy()

    # Criar coluna datetime
    try:
        df_plot['datetm'] = pd.to_datetime(df_plot[year_col].astype(str) + '-' +
                                           df_plot[mo_col].astype(str) + '-' +
                                           df_plot[dy_col].astype(str) + ' ' +
                                           df_plot[hr_col].astype(str).str.zfill(2) + ':00:00', errors='coerce')
    except Exception as e:
        print(f"‚ùå ERRO ao criar a coluna 'datetm' a partir das colunas de data/hora: {str(e)}")
        logging.error(f"Erro ao criar 'datetm' para NASA: {str(e)}")
        print('-' * 100)
        return

    df_plot.dropna(subset=['datetm', data_column], inplace=True) # Remove linhas onde datetm ou data_column s√£o NaT/NaN

    if df_plot.empty:
        print(f"‚ö†Ô∏è AVISO: DataFrame NASA ficou vazio ap√≥s cria√ß√£o de 'datetm' e remo√ß√£o de NAs. N√£o foi poss√≠vel gerar o gr√°fico.")
        logging.warning(f"Gr√°fico de Dispers√£o NASA n√£o gerado: DataFrame vazio ap√≥s processamento de 'datetm'.")
        print('-' * 100)
        return

    data_series = df_plot[data_column]
    time_series = df_plot['datetm']

    plt.figure(figsize=(12, 5))
    plt.plot(time_series, data_series, label=f'{data_column} {raw_data_label_suffix}', color=raw_data_color, alpha=0.5, linestyle='-', marker=None, linewidth=0.8)

    time_numeric = mdates.date2num(time_series)
    sns.regplot(
        x=time_numeric,
        y=data_series,
        scatter=False,
        lowess=True,
        color=trend_color,
        label='Tend√™ncia (LOESS)',
        line_kws={'linewidth': 2}
    )

    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=12))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45, ha='right')

    plt.title(f'{title_prefix} - {data_column} (01/2006 a 04/2007)')
    plt.xlabel(xlabel_text)
    plt.ylabel(ylabel_text)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"‚òë Gr√°fico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gr√°fico de Dispers√£o Temporal para NASA '{data_column}' salvo em '{output_filename}'.")
    print('-' * 100)

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ci√™ncia de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns: # Necess√°rio para o data_column
         dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')
    print("‚òë DataFrame NASA carregado e coluna WS50M verificada/convertida para num√©rico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gr√°ficos.")
except FileNotFoundError:
    print(f"‚ùå ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' n√£o encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"‚ùå ERRO ao carregar ou pr√©-processar dfNASA_jan2006_abr2007: {str(e)}. Gr√°ficos NASA n√£o gerados.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    plot_temporal_dispersion_nasa(
        dfNASA_jan2006_abr2007,
        year_col='YEAR', # Coluna de ano
        mo_col='MO',     # Coluna de m√™s
        dy_col='DY',     # Coluna de dia
        hr_col='HR',     # Coluna de hora
        data_column='WS50M',
        title_prefix='Dispers√£o Temporal (NASA)',
        xlabel_text='Tempo',
        ylabel_text='Velocidade do Vento (m/s) a 50m',
        output_filename='./dispersao_temporal_NASAWS50_melhorado.png'
    )
else:
    print("Skipping NASA dispersion plots due to empty or problematic DataFrame.")

print("--- Gera√ß√£o de Gr√°ficos de Dispers√£o Temporal para NASA POWER Conclu√≠da. ---")
print("-" * 100)
````

### üî® **3.1. Transformar Dataframe da REDE SONDA (01/2006-04/2007)**

Os dados da REDE SONDA, originalmente registrados a cada 10 minutos, foram reamostrados para frequ√™ncia **hor√°ria**. Para tornar os dados mais robustos:

- Foi realizada a **convers√£o da coluna de data/hora** (`datetm`) para o tipo `datetime`.
- Aplicada **reamostragem hor√°ria** com dois m√©todos de agrega√ß√£o:
  - **M√©dia** (`mean`)
  - **Mediana** (`median`)
- Gerada uma **m√©dia suavizada**, combinando m√©dia e mediana:
  
$$
\text{M√©dia Suavizada} = \frac{\text{M√©dia} + \text{Mediana}}{2}
$$

Essa abordagem reduz o impacto de outliers e flutua√ß√µes abruptas, proporcionando uma melhor estimativa do comportamento hor√°rio dos dados de vento.

````python
dfSonda_jan2006_abr2007 = pd.read_csv('/content/drive/MyDrive/Disciplinas/CieÃÇncia de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007/dfSonda_jan2006_abr2007.csv', sep=';')

dfNASA_jan2006_abr2007 = pd.read_csv('/content/drive/MyDrive/Disciplinas/CieÃÇncia de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv')


# Reamostragem dos dados da REDE SONDA de 10 em 10 minutos para 1 hora
print("DATAFRAME SONDA:\n\n", dfSonda_jan2006_abr2007.head(6))
print(f"\nDADOS NULOS SONDA:\n\n{dfSonda_jan2006_abr2007.isna().sum()}")
print("\n--------------------------------------------------------------------------\n")
print("DATAFRAME NASA:\n\n", dfNASA_jan2006_abr2007.head(6))
print(f"\nDADOS NULOS NASA:\n\n{dfNASA_jan2006_abr2007.isna().sum()}")
print("\n--------------------------------------------------------------------------\n")

# Convers√£o da coluna 'datetm' para datetime
dfSonda_jan2006_abr2007['datetm'] = pd.to_datetime(dfSonda_jan2006_abr2007['datetm'])

# Definir 'datetm' como √≠ndice para reamostragem
sonda_resample = dfSonda_jan2006_abr2007.set_index('datetm')

# Reamostragem hor√°ria com m√©dia e mediana
mean_df = sonda_resample.resample('1h').mean()
median_df = sonda_resample.resample('1h').median()

# C√°lculo da m√©dia suavizada
smoothed_df = (mean_df + median_df) / 2
smoothed_df = smoothed_df.reset_index()

# Adicionar colunas de data: ANO, MES, DIA, HORA
smoothed_df['ANO'] = smoothed_df['datetm'].dt.year
smoothed_df['MES'] = smoothed_df['datetm'].dt.month
smoothed_df['DIA'] = smoothed_df['datetm'].dt.day
smoothed_df['HORA'] = smoothed_df['datetm'].dt.hour

# Verificar colunas existentes
colunas_disponiveis = smoothed_df.columns.str.lower().tolist()
ws_col = next((col for col in colunas_disponiveis if 'ws' in col), None)
wd_col = next((col for col in colunas_disponiveis if 'wd' in col), None)

# Mapear nomes reais
ws_real = [col for col in smoothed_df.columns if col.lower() == ws_col][0]
wd_real = [col for col in smoothed_df.columns if col.lower() == wd_col][0]

# Selecionar e renomear colunas espec√≠ficas
df_jan2006_abr2007 = smoothed_df[['ANO', 'MES', 'DIA', 'HORA', ws_real, wd_real]].rename(columns={
    ws_real: 'SONDAWS50',
    wd_real: 'SONDAWD50'
})

# Visualiza√ß√£o final do nnovo DataFrame:
print("VERIFICA√á√ÉO INICIAL:")
print(f" - Dimens√£o do dataframe: {df_jan2006_abr2007.shape}")
print(f" - Dados duplicados     : {df_jan2006_abr2007.duplicated().sum()}")
print("\n--------------------------------------------------------------------------\n")
print(f"DADOS NULOS:\n\n{df_jan2006_abr2007.isna().sum()}")
print("\n--------------------------------------------------------------------------\n")
print(df_jan2006_abr2007.head(6).round(2))
print("\n--------------------------------------------------------------------------\n")
````

````python
# Agrupar e calcular m√©dias/medianas
df_grouped = (
    df_jan2006_abr2007.groupby(['ANO', 'MES'])
    .agg({
        'SONDAWS50': ['mean', 'median'],
        'NASAWS50': ['mean', 'median']
    })
    .reset_index()
)

# Simplificar colunas MultiIndex
df_grouped.columns = ['ANO', 'MES', 'SONDAWS50_mean', 'SONDAWS50_median', 'NASAWS50_mean', 'NASAWS50_median']

# Calcular m√©dia suavizada
df_grouped['SONDAWS50_svz'] = (df_grouped['SONDAWS50_mean'] + df_grouped['SONDAWS50_median']) / 2
df_grouped['NASAWS50_svz'] = (df_grouped['NASAWS50_mean'] + df_grouped['NASAWS50_median']) / 2

# Criar coluna MES/ANO
df_grouped['MES'] = df_grouped['MES'].astype(int).apply(lambda x: f"{x:02}")
df_grouped['ANO'] = df_grouped['ANO'].astype(str)
df_grouped['MES_ANO'] = df_grouped['MES'] + '/' + df_grouped['ANO']

# Ordenar
df_grouped = df_grouped.sort_values(['ANO', 'MES'])

# Exibir
print("-" * 100)
df_grouped.head().round(2)
print("-" * 100)
````

### üßÆ **C√°lculo do Erro Padr√£o entre SONDAWS50 e NASAWS50**
 O erro padr√£o da estimativa (SEE) mede a dispers√£o m√©dia dos dados em torno da linha de regress√£o.
  
 **F√≥rmula**:  

 $$
 \text{SEE} = \sqrt{\frac{\sum (y_i - \hat{y}_i)^2}{n - 2}}
 $$  

 Onde:  
 > - Valores reais de `NASAWS50`:<left>$$y_i$$<left>
 > - Valores previstos pela regress√£o linear: $$ \hat{y}_i $$  
 > - N√∫mero de observa√ß√µes: $$n$$

````python
from sklearn.linear_model import LinearRegression

# Supondo df_grouped j√° cont√©m as m√©dias suavizadas
X = df_grouped[['SONDAWS50_svz']]  # Vari√°vel independente
y = df_grouped['NASAWS50_svz']     # Vari√°vel dependente

# Ajustar modelo de regress√£o linear
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)  # Valores previstos

# Calcular res√≠duos e SEE
residuos = y - y_pred
soma_quadrados_residuos = np.sum(residuos**2)
n = len(y)
SEE = np.sqrt(soma_quadrados_residuos / (n - 2))

print(f'Erro Padr√£o da Estimativa (SEE):\n{SEE:.3f} m/s')
````

#### ‚úÖ **An√°lise de Resultados:**

- **`SEE = 0.563 m/s`**: Indica que, em m√©dia, os valores de NASAWS50 desviam-se **`¬±0.563m/s`** da linha de regress√£o ajustada.

- Quanto menor o **`SEE`**, mais precisa √© a rela√ß√£o linear. Neste caso, o valor sugere uma dispers√£o moderada, consistente com o **`R¬≤`** de **`0.47`**.

 O c√°lculo foi feito com m√©dias mensais suavizadas, o que pode subestimar a variabilidade real. Para maior precis√£o, utilize os dados brutos.

2. **Compara√ß√£o com Outras M√©tricas:**

 - O **`SEE`** complementa o **`R¬≤`**: Um **`R¬≤`** baixo e **`SEE`** alto refor√ßam a necessidade de investigar outros fatores explicativos.

 - No exemplo, **`47%`** da vari√¢ncia √© explicada, e o **`SEE`** quantifica o erro m√©dio absoluto.

base_url = r"https://power.larc.nasa.gov/api/temporal/daily/point?parameters=T2M,T2MDEW,T2MWET,TS,T2M_RANGE,T2M_MAX,T2M_MIN&community=RE&longitude={longitude}&latitude={latitude}&start=20150101&end=20150305&format=JSON"

    filename = response.headers['content-disposition'].split('filename=')[1]

    filepath = os.path.join(output, filename)
    with open(filepath, 'w') as file_object:
