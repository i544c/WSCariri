1. **INTRODUÇÃO:**

 ><b>Contexto Ambiental:</b></br>
     A energia eólica é uma das fontes renováveis mais promissoras para a diversificação da matriz energética brasileira, especialmente em regiões de clima semiárido, como o interior da Paraíba. No entanto, a análise e o planejamento de projetos eólicos dependem fortemente da disponibilidade de dados confiáveis sobre a velocidade e a direção dos ventos.</br>
 </br><b>Contexto Socioeconômico:</b></br>
     Outro aspecto importante é a análise de viabilidade econômica em projetos eólicos, que depende criticamente de dados precisos e contínuos sobre a velocidade do vento, pois permitem estimar o potencial energético e otimizar o retorno financeiro. No entanto, a escassez de dados confiáveis e padronizados dificulta o planejamento estratégico e a atração de investimentos.</br>
 </br><b>Desafios:</b></br>
     Atualmente, um dos principais desafios enfrentados por pesquisadores e formuladores de políticas públicas é a dificuldade de acesso a dados de velocidade do vento nacionais de qualidade. Os bancos de dados ambientais brasileiros são, em sua maioria, descentralizados,com formatos variados, lacunas temporais e dificuldades de integração. Além disso, muitas vezes esses dados estão organizados de forma semiestruturada, exigindo tratamento extensivo antes de sua aplicação em análises mais robustas.</br>
 </br><b>Objetivo:</b></br>
     Este projeto propõe o uso de dados diários de velocidade do vento (`50m altitude`) da estação da **Rede SONDA**, localizada em São José do Cariri – PB (`semiárido paraibano`), os quais se encontram distribuídos em arquivos mensais (`CSV`) e apresentam problemas como lacunas e inconsistências. Esses dados serão tratados e integrados a dados do projeto **NASA POWER DAV** (`utilizados como apoio`), com o objetivo de desenvolver modelos preditivos de velocidade do vento baseados em séries temporais.

 1.1. **Problema de ciência de dados:**

     O presente projeto enfrenta o desafio de construir um modelo preditivo da velocidade do vento com base em dados semiestruturados e incompletos provenientes da estação da Rede SONDA, localizada em São José do Cariri – PB.

 **Os principais problemas encontrados nesse banco de dados são:**

      * **`Fragmentação dos dados:`** os arquivos estão distribuídos mensalmente (um arquivo.csv por mês), exigindo concatenação e organização em uma única estrutura temporal.
      * **`Dados ausentes:`** ausência completa de registros para os meses de Maio/2007, Outubro, Novembro e Dezembro/2009.
      * **`Inconsistências temporais:`** possibilidade de sobreposição ou ausência de timestamps regulares entre arquivos.
      * **`Formatação heterogênea:`** diferenças em cabeçalhos, codificação, padrões de separação decimal e estrutura de colunas entre os arquivos.
      * **`Falta de metadados padronizados:`** ausência de documentação clara sobre unidades, padrões de qualidade e pré-processamentos já aplicados.

````python
# Algoritmo otimizado para extração das URLs da Rede Sonda:
print('=' * 80)
print("--- Iniciando a ETAPA 3: Geração de URLs da Rede Sonda ---")
print('=' * 80)

# Base da URL
BASE_URL = "https://sonda.ccst.inpe.br/basedados/dados/sonda/eolicos/SCR"

# Parâmetros de intervalo de datas
ano_inicio = 2006
ano_fim = 2007
limites_meses = {
    2006: 12,  # Janeiro a Dezembro
    2007: 4    # Janeiro a Abril
}

# Geração dinâmica de URLs com base no intervalo definido
urls_dados = []
for ano in range(ano_inicio, ano_fim + 1):
    for mes in range(1, limites_meses.get(ano, 12) + 1):
        aa = str(ano)[-2:]
        mm = f"{mes:02d}"
        # A extensão aqui é sempre '.zip' para este período (2006-2007)
        # O ".7z" é usado apenas para 2009 e será tratado em outro bloco de geração de URLs.
        url = f"{BASE_URL}/{ano}/SCR{aa}{mm}WD.zip"
        urls_dados.append(url)

# Verificação de resultado e exibição das URLs geradas
print(f"✅ {len(urls_dados)} URLs geradas para o período de {ano_inicio} a {ano_fim}:")
# Para evitar uma saída muito longa, exiba apenas as primeiras e últimas URLs, se houver muitas.
if len(urls_dados) > 10: # Ajuste este número conforme o desejado
    for i, url in enumerate(urls_dados):
        if i < 20 or i >= len(urls_dados) - 5: # Mostra as 5 primeiras e 5 últimas
            print(f"   - {url}")
        elif i == 20:
            print("   ...") # Indica que há mais URLs no meio
    print(f"\nDetalhes de todas as URLs geradas são registrados no log.")
else:
    for url in urls_dados:
        print(f"   - {url}")

logging.info(f"Total de URLs geradas: {len(urls_dados)}. Lista completa: {urls_dados}")

print('-' * 80)
print("--- ETAPA 3 Concluída: URLs geradas. ---")
print("-" * 80)
````

4. **Baixar Arquivos ZIP, Extrair CSVs e Salvar Individualmente (com verificação)**

````python
# Importar tqdm para a barra de progresso (adicione no início do seu script se ainda não estiver lá)
from tqdm.notebook import tqdm

# Etapa 4: Baixar Arquivos ZIP, Extrair CSVs e Salvar Individualmente (com verificação)
print('=' * 80)
print("--- Iniciando Etapa 4: Download e salvamento de arquivos CSV individuais ---")
print('=' * 80)

# Adicionando a barra de progresso com tqdm
for url in tqdm(urls_dados, desc="Processando arquivos da Rede SONDA"):
    nome_arquivo_origem = url.split('/')[-1]
    csv_filename_destino = nome_arquivo_origem.replace('.zip', '.csv').replace('.7z', '.csv') # Adicionado .replace('.7z', '.csv') para consistência com 2009
    csv_path_destino = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, csv_filename_destino)

    # Verificar se o arquivo CSV individual já existe
    if os.path.exists(csv_path_destino):
        try:
            df_existente_check = pd.read_csv(csv_path_destino, sep=";", header=None, nrows=1, encoding="latin1")
            if df_existente_check.shape[1] != len(colunas):
                print(f"🔄 Arquivo '{csv_filename_destino}' com formato inválido. Removendo e reprocessando.")
                logging.warning(f"Arquivo '{csv_filename_destino}' ({df_existente_check.shape[1]} cols) com formato de colunas diferente do esperado ({len(colunas)} cols). Removendo para reprocessamento.")
                os.remove(csv_path_destino)
            else:
                print(f"⏩ Arquivo '{csv_filename_destino}' já existe e tem formato correto. Pulando download.")
                logging.info(f"Arquivo '{csv_filename_destino}' já existe e tem o formato correto ({len(colunas)} cols). Pulando download e processamento.")
                continue
        except pd.errors.EmptyDataError:
            print(f"🗑️ Arquivo '{csv_filename_destino}' existe mas está vazio. Removendo para reprocessar.")
            logging.warning(f"Arquivo '{csv_filename_destino}' existe mas está vazio. Removendo para reprocessamento.")
            os.remove(csv_path_destino)
        except Exception as e:
            print(f"❌ Erro ao verificar '{csv_filename_destino}': {e}. Removendo para reprocessar.")
            logging.error(f"Erro ao verificar arquivo existente '{csv_filename_destino}': {e}. Removendo para tentar reprocessamento.")
            try:
                os.remove(csv_path_destino)
            except OSError as oe:
                print(f"Não foi possível remover '{csv_filename_destino}': {oe}")
                logging.error(f"Não foi possível remover o arquivo '{csv_filename_destino}' para reprocessamento: {oe}")

    try:
        # print(f"Baixando e processando '{url}'...") # Pode manter ou remover, dependendo do quão detalhado quer o output.
        logging.info(f"Baixando e processando '{url}'...")
        response = requests.get(url, timeout=30) # Adicionado timeout para evitar travamentos indefinidos
        response.raise_for_status() # Levanta um erro HTTP para códigos de status ruins (4XX ou 5XX)

        extensao_arquivo = nome_arquivo_origem.split('.')[-1].lower()
        df_processado = None

        if extensao_arquivo == 'zip':
            # Abrir o arquivo ZIP a partir do conteúdo baixado em memória
            with zipfile.ZipFile(BytesIO(response.content)) as z:
                # Listar todos os arquivos dentro do ZIP
                nomes_csv_no_zip = [f_name for f_name in z.namelist() if f_name.lower().endswith(".csv")] # Considerar .csv em minúsculas
                if not nomes_csv_no_zip:
                    print(f"⚠️ Nenhum arquivo .csv encontrado no ZIP: {url}")
                    logging.warning(f"Nenhum arquivo .csv encontrado dentro do ZIP: {url}")
                    continue
                nome_csv_principal_zip = nomes_csv_no_zip[0]
                logging.info(f"Encontrado arquivo CSV '{nome_csv_principal_zip}' dentro do ZIP.")
                # Abrir o arquivo CSV de dentro do ZIP
                with z.open(nome_csv_principal_zip) as f_csv:
                    temp_df = pd.read_csv(f_csv, sep=";", header=None, encoding="latin1", low_memory=False)
                    # Certifique-se de que a atribuição a df_processado ocorra aqui para o bloco de checagem de colunas
                    df_processado = temp_df

        elif extensao_arquivo == '7z':
            temp_extract_dir = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, f"temp_extract_{nome_arquivo_origem}")
            os.makedirs(temp_extract_dir, exist_ok=True)
            logging.info(f"Criado diretório temporário para 7z: {temp_extract_dir}")
            try:
                with py7zr.SevenZipFile(BytesIO(response.content), mode='r') as archive:
                    nomes_arquivos_no_7z = archive.getnames()
                    nomes_csv_no_7z = [f_name for f_name in nomes_arquivos_no_7z if f_name.lower().endswith('.csv')]
                    if not nomes_csv_no_7z:
                        print(f"⚠️ Nenhum arquivo .csv encontrado dentro do 7z: {url}")
                        logging.warning(f"Nenhum arquivo .csv encontrado dentro do 7z: {url}")
                        continue
                    nome_csv_principal_7z = nomes_csv_no_7z[0]
                    logging.info(f"Extraindo '{nome_csv_principal_7z}' de dentro do 7z '{nome_arquivo_origem}' para {temp_extract_dir}")
                    archive.extract(path=temp_extract_dir, targets=[nome_csv_principal_7z])
                    caminho_csv_extraido_temp = os.path.join(temp_extract_dir, nome_csv_principal_7z)
                    temp_df = pd.read_csv(caminho_csv_extraido_temp, sep=";", header=None, encoding="latin1", low_memory=False)
                    # Certifique-se de que a atribuição a df_processado ocorra aqui
                    df_processado = temp_df
            finally:
                if os.path.exists(temp_extract_dir):
                    shutil.rmtree(temp_extract_dir)
                    logging.info(f"Diretório temporário {temp_extract_dir} removido.")
        else:
            print(f"⚠️ Extensão de arquivo não suportada '{extensao_arquivo}' para {url}")
            logging.warning(f"Extensão de arquivo não suportada '{extensao_arquivo}' para {url}")
            continue

        if df_processado is not None:
            # Correção para o caso de arquivos de 2009 com 12 colunas, remover a última
            # Esta lógica deve ser aplicada ao df_processado, que é o que contém os dados lidos
            if "09" in csv_filename_destino and df_processado.shape[1] == 12:
                df_processado = df_processado.iloc[:, :-1]
                print(f"⚙️ Coluna extra removida de '{csv_filename_destino}' (original: {nome_arquivo_origem}).")
                logging.info(f"Última coluna removida do arquivo '{csv_filename_destino}' (original: {nome_arquivo_origem}), novo shape: {df_processado.shape}")

            if df_processado.shape[1] == len(colunas):
                df_processado.to_csv(target_csv_path, index=False, header=False, sep=";", encoding="latin1")
                print(f"✅ Arquivo salvo: '{csv_filename_destino}' em '{target_csv_path}'.") # Mensagem de sucesso com caminho
                logging.info(f"Arquivo processado e salvo como: '{target_csv_path}' com {df_processado.shape[0]} linhas e {df_processado.shape[1]} colunas.")
            else:
                print(f"❌ ERRO: '{csv_filename_destino}' (de {nome_arquivo_origem}) tem {df_processado.shape[1]} cols, esperado {len(colunas)}. NÃO SALVO.")
                logging.error(f"Após processamento, arquivo '{csv_filename_destino}' (de {nome_arquivo_origem}) possui {df_processado.shape[1]} colunas, mas o esperado era {len(colunas)}. Arquivo não salvo.")
        else:
            print(f"⚠️ Nenhum DataFrame foi processado para a URL {url}")
            logging.warning(f"Nenhum DataFrame foi processado para a URL {url}.")

    except requests.exceptions.RequestException as e:
        print(f"❌ Erro de requisição ao baixar {url}: {e}")
        logging.error(f"Erro de requisição ao baixar {url}: {e}")
    except (zipfile.BadZipFile, py7zr.exceptions.Bad7zFile) as e:
        print(f"❌ Erro: Arquivo baixado de {url} corrompido ou inválido.")
        logging.error(f"Erro: Arquivo baixado de {url} (ZIP/7Z) não é válido ou está corrompido: {e}")
    except Exception as e:
        print(f"❌ ERRO INESPERADO ao processar {url}: {str(e)}")
        logging.error(f"Erro inesperado ao processar {url}: {str(e)}")

print("-" * 80)
print("Etapa 4 concluída!!!")
print("-" * 80)
````

````python
# --- ETAPA 1: DEFINIR CAMINHOS DE DESTINO ---
import os
import pandas as pd
import requests
import zipfile
import shutil
from io import BytesIO
import logging
from IPython.display import display # Para exibir DataFrames de forma mais elegante no Colab
from tqdm.notebook import tqdm # Importando tqdm para barras de progresso no Jupyter/Colab

# Configurar logging (útil para um rastreamento mais detalhado se necessário)
# Se esta célula for executada múltiplas vezes, basicConfig pode não ter efeito após a primeira vez
# sem force=True (Python 3.8+), mas as chamadas logging.info() ainda funcionarão.
if not logging.getLogger().handlers: # Configura apenas se nenhum handler estiver configurado
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Instalação automática de py7zr, se necessário
try:
    import py7zr
    logging.info("py7zr já está instalado.") # Mensagem de log, não print
except ImportError:
    logging.info("py7zr não encontrado. Tentando instalar...") # Mensagem de log
    import subprocess
    import sys
    try:
        print("Instalando py7zr... Aguarde.") # Mensagem para o usuário
        subprocess.check_call([sys.executable, "-m", "pip", "install", "py7zr"])
        import py7zr
        print("py7zr instalado com sucesso.")
        logging.info("py7zr instalado com sucesso via pip.")
    except Exception as e:
        print(f"ERRO: Falha ao instalar py7zr: {e}")
        logging.error(f"Falha ao instalar py7zr: {e}")
        raise ImportError("py7zr não pôde ser instalado. O script não pode continuar sem ele para arquivos .7z.")

print('=' *80)
print("--- Iniciando a ETAPA 1: Definições Iniciais e Configurações ---")
print('=' *80)

# Configurações globais
BASE_DRIVE_PATH = '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina' # Ajuste se necessário
HEADER_FILE_PATH = os.path.join(BASE_DRIVE_PATH, 'WD_header.csv')
INDIVIDUAL_CSV_SAVE_PATH = os.path.join(BASE_DRIVE_PATH, 'DadosSonda_Individuais_Jun2007Set2009')
FINAL_DATASET_SAVE_PATH = os.path.join(BASE_DRIVE_PATH, 'DadosSonda_Unificados_Jun2007Set2009')
FINAL_DATASET_NAME = 'dfSonda_jun2007_set2009.csv'

# Montar Google Drive (se ainda não montado) - geralmente feito em célula separada
# from google.colab import drive
# drive.mount('/content/drive')
# print("\nGoogle Drive Montado (ou já estava montado).") # Adicionar após a montagem real

os.makedirs(INDIVIDUAL_CSV_SAVE_PATH, exist_ok=True)
os.makedirs(FINAL_DATASET_SAVE_PATH, exist_ok=True)

print(f"\nCaminho do arquivo de cabeçalho: \n{HEADER_FILE_PATH}")
print(f"\nPasta para CSVs individuais: \n{INDIVIDUAL_CSV_SAVE_PATH}")
print(f"\nPasta para dataset final: \n{FINAL_DATASET_SAVE_PATH}")
print('-' *80)
print("ETAPA 1 Concluída: Caminhos definidos e pastas criadas/verificadas.\n")

# --- ETAPA 2: CARREGAR CABEÇALHO DE REFERÊNCIA ---
print('=' * 80)
print("--- Iniciando a ETAPA 2: Carregar Cabeçalho de Referência ---")
print('=' *80)
colunas = []
if not os.path.exists(HEADER_FILE_PATH):
    print(f"ERRO FATAL: Arquivo de cabeçalho '{HEADER_FILE_PATH}' não encontrado. Verifique o caminho.")
    logging.error(f"Arquivo de cabeçalho '{HEADER_FILE_PATH}' não encontrado.")
    raise FileNotFoundError(f"Arquivo de cabeçalho '{HEADER_FILE_PATH}' não encontrado.")
else:
    header_df = pd.read_csv(HEADER_FILE_PATH, header=None, sep=";")
    colunas = header_df.iloc[0].tolist()
    print(f"Cabeçalhos carregados com sucesso. Número de colunas esperado: {len(colunas)}")
    logging.info(f"Cabeçalhos carregados: {colunas} | Número de colunas: {len(colunas)}")
print('-' *80)
print("ETAPA 2 Concluída: Cabeçalho carregado.\n")

# --- ETAPA 3: GERAÇÃO DE URLs ---
print('=' * 80)
print("--- Iniciando a ETAPA 3: Geração de URLs da Rede Sonda ---")
print('=' *80)
BASE_URL = "https://sonda.ccst.inpe.br/basedados/dados/sonda/eolicos/SCR"
ano_inicio = 2007
ano_fim = 2009
limites_meses = {2007: 12, 2008: 12, 2009: 9}
urls_dados = []

for ano in range(ano_inicio, ano_fim + 1):
    inicio_mes = 6 if ano == 2007 else 1
    fim_mes_ano = limites_meses.get(ano, 12)
    for mes in range(inicio_mes, fim_mes_ano + 1):
        aa = str(ano)[-2:]
        mm = f"{mes:02d}"
        ext = '7z' if ano == 2009 else 'zip'
        url = f"{BASE_URL}/{ano}/SCR{aa}{mm}WD.{ext}"
        urls_dados.append(url)

print(f"Total de {len(urls_dados)} URLs geradas para o período de Jun/2007 a Set/2009.")
logging.info(f"Total de URLs geradas: {len(urls_dados)}")
print('-' * 80)
print("ETAPA 3 Concluída: URLs geradas.\n")

# --- ETAPA 4: DOWNLOAD, EXTRAÇÃO E SALVAMENTO CONDICIONAL (COM TRATAMENTO DE COLUNAS) ---
print('=' * 80)
print("--- Iniciando a ETAPA 4: Download, Extração e Salvamento Condicional ---")
print('=' *80)
# Adicionando a barra de progresso com tqdm
for url in tqdm(urls_dados, desc="Processando arquivos da Rede SONDA"):
    nome_arquivo_origem = url.split('/')[-1]
    csv_filename_destino = nome_arquivo_origem.replace('.zip', '.csv').replace('.7z', '.csv')
    csv_path_destino = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, csv_filename_destino)

    if os.path.exists(csv_path_destino):
        try:
            df_existente_check = pd.read_csv(csv_path_destino, sep=";", header=None, nrows=1, encoding="latin1")
            if df_existente_check.shape[1] != len(colunas):
                print(f"Arquivo '{csv_filename_destino}' ({df_existente_check.shape[1]} cols) com formato inválido. Removendo para reprocessar.")
                logging.warning(f"Arquivo '{csv_filename_destino}' ({df_existente_check.shape[1]} cols) com formato de colunas diferente do esperado ({len(colunas)} cols). Removendo para reprocessamento.")
                os.remove(csv_path_destino)
            else:
                # print(f"Arquivo '{csv_filename_destino}' já existe e tem formato correto. Pulando.") # Removido para não poluir o output com tqdm
                logging.info(f"Arquivo '{csv_filename_destino}' já existe e tem o formato correto ({len(colunas)} cols). Pulando download e processamento.")
                continue
        except pd.errors.EmptyDataError:
            print(f"Arquivo '{csv_filename_destino}' existe mas está vazio. Removendo para reprocessar.")
            logging.warning(f"Arquivo '{csv_filename_destino}' existe mas está vazio. Removendo para reprocessamento.")
            os.remove(csv_path_destino)
        except Exception as e:
            print(f"Erro ao verificar '{csv_filename_destino}': {e}. Removendo para reprocessar.")
            logging.error(f"Erro ao verificar arquivo existente '{csv_filename_destino}': {e}. Removendo para tentar reprocessamento.")
            try:
                os.remove(csv_path_destino)
            except OSError as oe:
                print(f"Não foi possível remover '{csv_filename_destino}': {oe}")
                logging.error(f"Não foi possível remover o arquivo '{csv_filename_destino}' para reprocessamento: {oe}")
    try:
        # print(f"Tentando baixar e processar: {url}") # Removido para não poluir o output com tqdm
        logging.info(f"Baixando arquivo de: {url}")
        response = requests.get(url, timeout=60)
        response.raise_for_status()

        extensao_arquivo = nome_arquivo_origem.split('.')[-1].lower()
        df_processado = None

        if extensao_arquivo == 'zip':
            with zipfile.ZipFile(BytesIO(response.content)) as z:
                nomes_csv_no_zip = [f_name for f_name in z.namelist() if f_name.lower().endswith('.csv')]
                if not nomes_csv_no_zip:
                    print(f"AVISO: Nenhum .csv encontrado dentro do ZIP: {url}")
                    logging.warning(f"Nenhum arquivo .csv encontrado dentro do ZIP: {url}")
                    continue
                nome_csv_principal_zip = nomes_csv_no_zip[0]
                # print(f"  Processando '{nome_csv_principal_zip}' do ZIP '{nome_arquivo_origem}'...") # Removido
                logging.info(f"Processando '{nome_csv_principal_zip}' de dentro do ZIP '{nome_arquivo_origem}'")
                with z.open(nome_csv_principal_zip) as f:
                    df_processado = pd.read_csv(f, sep=";", header=None, encoding="latin1", low_memory=False)

        elif extensao_arquivo == '7z':
            temp_extract_dir = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, f"temp_extract_{nome_arquivo_origem}")
            os.makedirs(temp_extract_dir, exist_ok=True)
            logging.info(f"Criado diretório temporário para 7z: {temp_extract_dir}")
            try:
                with py7zr.SevenZipFile(BytesIO(response.content), mode='r') as archive:
                    nomes_arquivos_no_7z = archive.getnames()
                    nomes_csv_no_7z = [f_name for f_name in nomes_arquivos_no_7z if f_name.lower().endswith('.csv')]
                    if not nomes_csv_no_7z:
                        print(f"AVISO: Nenhum .csv encontrado dentro do 7z: {url}")
                        logging.warning(f"Nenhum arquivo .csv encontrado dentro do 7z: {url}")
                        continue
                    nome_csv_principal_7z = nomes_csv_no_7z[0]
                    # print(f"  Extraindo e processando '{nome_csv_principal_7z}' do 7z '{nome_arquivo_origem}'...") # Removido
                    logging.info(f"Extraindo '{nome_csv_principal_7z}' de dentro do 7z '{nome_arquivo_origem}' para {temp_extract_dir}")
                    archive.extract(path=temp_extract_dir, targets=[nome_csv_principal_7z])
                    caminho_csv_extraido_temp = os.path.join(temp_extract_dir, nome_csv_principal_7z)
                    df_processado = pd.read_csv(caminho_csv_extraido_temp, sep=";", header=None, encoding="latin1", low_memory=False)
            finally:
                if os.path.exists(temp_extract_dir):
                    shutil.rmtree(temp_extract_dir)
                    logging.info(f"Diretório temporário {temp_extract_dir} removido.")
        else:
            print(f"AVISO: Extensão de arquivo não suportada '{extensao_arquivo}' para {url}")
            logging.warning(f"Extensão de arquivo não suportada '{extensao_arquivo}' para {url}")
            continue

        if df_processado is not None:
            if "09" in csv_filename_destino and df_processado.shape[1] == 12:
                df_processado = df_processado.iloc[:, :-1]
                print(f"  Coluna extra removida de '{csv_filename_destino}' (original: {nome_arquivo_origem}).")
                logging.info(f"Última coluna removida do arquivo '{csv_filename_destino}' (original: {nome_arquivo_origem}), novo shape: {df_processado.shape}")

            if df_processado.shape[1] == len(colunas):
                df_processado.to_csv(csv_path_destino, index=False, header=False, sep=";", encoding="latin1")
                # print(f"  Arquivo salvo: '{csv_path_destino}' ({df_processado.shape[0]} linhas, {df_processado.shape[1]} colunas).") # Removido
                logging.info(f"Arquivo processado e salvo como: '{csv_path_destino}' com {df_processado.shape[0]} linhas e {df_processado.shape[1]} colunas.")
            else:
                print(f"ERRO: Após processamento, '{csv_filename_destino}' (de {nome_arquivo_origem}) tem {df_processado.shape[1]} cols, esperado {len(colunas)}. NÃO SALVO.")
                logging.error(f"Após processamento, arquivo '{csv_filename_destino}' (de {nome_arquivo_origem}) possui {df_processado.shape[1]} colunas, mas o esperado era {len(colunas)}. Arquivo não salvo.")
        else:
            print(f"AVISO: Nenhum DataFrame foi processado para a URL {url}")
            logging.warning(f"Nenhum DataFrame foi processado para a URL {url}.")

    except requests.exceptions.RequestException as e:
        print(f"ERRO de requisição ao baixar {url}: {e}")
        logging.error(f"Erro de requisição ao baixar {url}: {e}")
    except zipfile.BadZipFile:
        print(f"ERRO: Arquivo ZIP de {url} inválido/corrompido.")
        logging.error(f"Erro: Arquivo baixado de {url} (ZIP) não é válido ou está corrompido.")
    except py7zr.exceptions.Bad7zFile:
        print(f"ERRO: Arquivo 7Z de {url} inválido/corrompido.")
        logging.error(f"Erro: Arquivo baixado de {url} (7Z) não é válido ou está corrompido.")
    except Exception as e:
        print(f"ERRO INESPERADO ao processar {url}: {str(e)}")
        logging.error(f"Erro inesperado ao processar {url}: {str(e)}")
print('-' * 80)
print("ETAPA 4 Concluída!!!\n")

# --- ETAPA 5: UNIFICAÇÃO DOS ARQUIVOS CSV INDIVIDUAIS (COM VERIFICAÇÃO) ---
print('=' * 80)
print("--- Iniciando a ETAPA 5: Unificação dos Arquivos CSV ---")
print('=' * 80)
arquivos_csv_individuais = sorted([f for f in os.listdir(INDIVIDUAL_CSV_SAVE_PATH) if f.lower().endswith('.csv') and os.path.isfile(os.path.join(INDIVIDUAL_CSV_SAVE_PATH, f))])
lista_dataframes_para_unir = []

if not arquivos_csv_individuais:
    print("AVISO: Nenhum arquivo CSV individual encontrado para unificação.")
    logging.warning("Nenhum arquivo CSV individual encontrado para unificação.")
else:
    print(f"Encontrados {len(arquivos_csv_individuais)} arquivos CSV para tentar unificar.")
    # Adicionando a barra de progresso com tqdm
    for nome_arquivo_csv in tqdm(arquivos_csv_individuais, desc="Unificando arquivos CSV"):
        caminho_completo_csv = os.path.join(INDIVIDUAL_CSV_SAVE_PATH, nome_arquivo_csv)
        try:
            df_temp = pd.read_csv(caminho_completo_csv, sep=";", header=None, encoding="latin1", low_memory=False)
            if df_temp.shape[1] == len(colunas):
                lista_dataframes_para_unir.append(df_temp)
                # print(f"Lido arquivo '{nome_arquivo_csv}' para unificação ({df_temp.shape[0]} linhas).") # Removido
                logging.info(f"Arquivo '{nome_arquivo_csv}' ({df_temp.shape[0]} linhas, {df_temp.shape[1]} colunas) carregado para unificação.")
            else:
                print(f"ERRO: Arquivo '{nome_arquivo_csv}' tem {df_temp.shape[1]} cols, esperado {len(colunas)}. IGNORADO.")
                logging.error(f"Arquivo '{nome_arquivo_csv}' tem {df_temp.shape[1]} colunas, mas o esperado era {len(colunas)}. Ignorando este arquivo na unificação.")
        except pd.errors.EmptyDataError:
            print(f"AVISO: Arquivo '{nome_arquivo_csv}' está vazio. IGNORADO.")
            logging.warning(f"Arquivo '{nome_arquivo_csv}' está vazio. Ignorando.")
        except Exception as e:
            print(f"ERRO ao ler '{nome_arquivo_csv}': {str(e)}. IGNORADO.")
            logging.error(f"Erro ao ler o arquivo '{nome_arquivo_csv}' para unificação: {str(e)}")

dataset_unificado = pd.DataFrame()
if lista_dataframes_para_unir:
    dataset_unificado = pd.concat(lista_dataframes_para_unir, ignore_index=True)
    print(f"\nTotal de {len(lista_dataframes_para_unir)} arquivos CSV unificados com sucesso.")
    print(f"Dimensões do dataset unificado ANTES de adicionar cabeçalho: {dataset_unificado.shape}")
    logging.info(f"Dataset unificado criado com sucesso: {dataset_unificado.shape[0]} linhas x {dataset_unificado.shape[1]} colunas.")
else:
    print("\nNenhum DataFrame válido foi carregado, o dataset unificado está vazio.")
    logging.warning("Nenhum DataFrame válido foi carregado, o dataset unificado está vazio.")
print('-' * 80)
print("ETAPA 5 Concluída!!!\n")

# --- ETAPA 6: APLICAR CABEÇALHO, EXIBIR AMOSTRA E SALVAR ---
print('=' * 80)
print("--- Iniciando a ETAPA 6: Aplicar Cabeçalho, Exibir Amostra e Salvar Dataset Final ---")
print('=' * 80)
if not dataset_unificado.empty:
    if len(colunas) == dataset_unificado.shape[1]:
        try:
            dataset_unificado.columns = colunas
            print("Cabeçalho adicionado ao dataset unificado!")
            logging.info("Cabeçalho aplicado ao dataset unificado.")

            print("-" *80)
            print("Primeiras linhas do dataset com cabeçalho:")
            print("-" *80)
            display(dataset_unificado.head(1).round(2)) # Usando display para melhor formatação

            num_duplicados = dataset_unificado.duplicated().sum()
            print("-" *80)
            print("\nCARACTERÍSTICAS DO DATAFRAME:")
            print(f" - Dimensão do dataframe: {dataset_unificado.shape}")
            print(f" - Dados duplicados      : {num_duplicados}")
            print("-" *80)
            print('\nDADOS NULOS:')
            print(dataset_unificado[['ws_50', 'wd_50']].isna().sum())
            print("-" *80)

            caminho_dataset_final = os.path.join(FINAL_DATASET_SAVE_PATH, FINAL_DATASET_NAME)
            if os.path.exists(caminho_dataset_final):
                print(f"\nDataset final '{FINAL_DATASET_NAME}' já existe em:\n'{FINAL_DATASET_SAVE_PATH}'.")
                print("Nenhum novo salvamento será realizado.")
                logging.info(f"Dataset final '{FINAL_DATASET_NAME}' já existe em '{FINAL_DATASET_SAVE_PATH}'. Pulando salvamento.")
            else:
                dataset_unificado.to_csv(caminho_dataset_final, index=False, sep=";", encoding="latin1")
                print(f"\nDataset final salvo com sucesso em:\n'{caminho_dataset_final}'")
                logging.info(f"Dataset final salvo com sucesso em: '{caminho_dataset_final}'")
        except Exception as e:
            print(f"ERRO ao aplicar cabeçalho ou salvar o dataset final: {str(e)}")
            logging.error(f"Erro ao aplicar cabeçalho ou salvar o dataset final: {str(e)}")
            logging.error(f"Detalhes - Shape do dataset unificado: {dataset_unificado.shape}, Número de cabeçalhos: {len(colunas)}")
    else:
        print(f"ERRO FATAL: Falha ao aplicar cabeçalho. Número de colunas no cabeçalho ({len(colunas)}) não corresponde ao dataset unificado ({dataset_unificado.shape[1]}).")
        logging.error(f"Falha ao aplicar cabeçalho: Número de colunas no cabeçalho ({len(colunas)}) não corresponde ao dataset unificado ({dataset_unificado.shape[1]}).")
else:
    print("AVISO: Dataset unificado está vazio. Nenhuma ação de salvamento ou aplicação de cabeçalho realizada.")
    logging.warning("Dataset unificado está vazio. Nenhuma ação de salvamento ou aplicação de cabeçalho realizada.")

print('-' * 80)
print("ETAPA 6 Concluída. PROCESSO FINALIZADO!!!")
print("-" * 80)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado

# --- Função Auxiliar (Repetida para Independência) ---
def plot_boxplot_with_outliers_info_sonda(dataframe, column, title_prefix, xlabel_text, output_filename, color='orange'):
    """
    Gera um boxplot para uma coluna específica do DataFrame da Rede SONDA,
    incluindo informações de outliers (IQR) e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o boxplot.
        title_prefix (str): Prefixo para o título do gráfico (ex: "Boxplot -").
        xlabel_text (str): Rótulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        color (str, optional): Cor do boxplot. Padrão é 'orange'.
    """
    print('=' * 100)
    print(f"--- Gerando Boxplot para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame SONDA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o boxplot.")
        logging.warning(f"Boxplot SONDA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"⚠️ AVISO: A série SONDA '{column}' está vazia após remover NAs. Não foi possível gerar o boxplot.")
        logging.warning(f"Boxplot SONDA não gerado: Série '{column}' vazia após dropna.")
        print('-' * 100)
        return

    Q1 = data_series.quantile(0.25)
    Q3 = data_series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound_iqr = Q1 - 1.5 * IQR
    upper_bound_iqr = Q3 + 1.5 * IQR

    outliers_count = data_series[(data_series < lower_bound_iqr) | (data_series > upper_bound_iqr)].count()
    total_count = len(data_series)
    outliers_percentage = (outliers_count / total_count * 100) if total_count > 0 else 0

    plt.figure(figsize=(9, 3))
    sns.boxplot(x=data_series, color=color, linewidth=1.5, flierprops=dict(marker='o', markerfacecolor='red', markersize=5))

    plt.annotate(
        f'Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\nOutliers: {outliers_count} ({outliers_percentage:.2f}%)',
        xy=(0.95, 0.05),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='bottom'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Boxplot para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Aplicação para dfSonda_jan2006_abr2007 ---
print('=' * 100)
print("--- Gerando gráficos para dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)

# O DataFrame dfSonda_jan2006_abr2007 deve estar carregado e pré-processado nas células anteriores.
# Exemplo de como garantir que ele esteja carregado, se esta célula for executada isoladamente:
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")
    # Garantir conversão de N/A e N/S para NaN (fundamental para as métricas)
    for col in ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace({'N/A': np.nan, 'N/S': np.nan}).astype(float)
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfSonda_jan2006_abr2007: {str(e)}. Gráficos SONDA não gerados.")
    logging.error(f"Erro no setup para gráficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame() # Define como vazio para pular gráficos

if not dfSonda_jan2006_abr2007.empty:
    # Gerar boxplot para ws_50 (SONDA)
    plot_boxplot_with_outliers_info_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50',
        'Boxplot - Velocidade do Vento 50m (SONDA)',
        'Velocidade (m/s)',
        './boxplot_SondaWS50.png',
        color='orange'
    )

    # Gerar boxplot para wd_50 (SONDA)
    plot_boxplot_with_outliers_info_sonda(
        dfSonda_jan2006_abr2007,
        'wd_50',
        'Boxplot - Direção do Vento 50m (SONDA)',
        'Direção (°)',
        './boxplot_SondaWD50.png',
        color='lightcoral'
    )
else:
    print("Skipping SONDA boxplots due to empty or problematic DataFrame.")

print("--- Geração de Boxplots para Rede SONDA Concluída. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado
from scipy.stats import norm # Para a curva de distribuição normal

# Configurar logging (garanta que só é configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Densidade ---
def plot_density_with_normal_sonda(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='skyblue', color_normal='red'):
    """
    Gera um gráfico de densidade para uma coluna específica do DataFrame da Rede SONDA,
    incluindo uma curva de distribuição normal teórica e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o gráfico de densidade.
        title_prefix (str): Prefixo para o título do gráfico (ex: "Distribuição de Densidade -").
        xlabel_text (str): Rótulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        color_density (str, optional): Cor da densidade empírica. Padrão é 'skyblue'.
        color_normal (str, optional): Cor da curva normal. Padrão é 'red'.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Densidade para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame SONDA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade SONDA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"⚠️ AVISO: A série SONDA '{column}' está vazia após remover NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade SONDA não gerado: Série '{column}' vazia após dropna.")
        print('-' * 100)
        return

    # Calcular média e desvio padrão para a curva normal
    media = data_series.mean()
    desvio = data_series.std()

    plt.figure(figsize=(8, 5)) # Ajustei o tamanho para melhor visualização
    sns.kdeplot(data_series, fill=True, label='Densidade Empírica', color=color_density, linewidth=2)

    # Gerar pontos para a curva normal teórica
    x_normal = np.linspace(data_series.min(), data_series.max(), 200)
    y_normal = norm.pdf(x_normal, media, desvio)
    plt.plot(x_normal, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribuição Normal Teórica')

    # Adicionar linha vertical para a média
    plt.axvline(media, color='black', linestyle=':', linewidth=1.5, label=f'Média: {media:.2f}')

    plt.annotate(
        f'Média: {media:.2f}\nDesvio Padrão: {desvio:.2f}',
        xy=(0.95, 0.85), # Ajustado para não sobrepor a legenda padrão
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade') # Adicionado rótulo do eixo Y
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Densidade para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Carregamento do DataFrame dfSonda_jan2006_abr2007 ---
# Esta parte é semelhante à sua função de boxplot para garantir que o dataframe esteja carregado
print('=' * 100)
print("--- Carregando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    # Garantir conversão de N/A e N/S para NaN e tipo float para colunas relevantes
    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            # Primeiro, substitui strings problemáticas por NaN
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan) # Adicionado -999 e -999.0 se forem usados para nulos
            # Tenta converter para float, os erros viram NaN
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')
    print("☑ DataFrame carregado e colunas relevantes convertidas para numérico.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gráficos.")

except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame() # Define como vazio para pular gráficos
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfSonda_jan2006_abr2007: {str(e)}. Gráficos SONDA não gerados.")
    logging.error(f"Erro no setup para gráficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gráfico de densidade para ws_50 (SONDA)
    plot_density_with_normal_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50',
        'Distribuição de Densidade (SONDA)',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_normal_SondaWS50_melhorado.png',
        color_density='deepskyblue', # Cor alterada para exemplo
        color_normal='crimson'      # Cor alterada para exemplo
    )
else:
    print("Skipping SONDA density plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Densidade para Rede SONDA Concluída. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado
from scipy.stats import weibull_min # Para a curva de distribuição Weibull

# Configurar logging (garanta que só é configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Densidade com Weibull ---
def plot_density_with_weibull_sonda(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='skyblue', color_weibull='green'):
    """
    Gera um gráfico de densidade para uma coluna específica do DataFrame da Rede SONDA,
    incluindo uma curva de distribuição Weibull teórica e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o gráfico de densidade.
        title_prefix (str): Prefixo para o título do gráfico (ex: "Distribuição de Densidade -").
        xlabel_text (str): Rótulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        color_density (str, optional): Cor da densidade empírica. Padrão é 'skyblue'.
        color_weibull (str, optional): Cor da curva Weibull. Padrão é 'green'.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Densidade com Weibull para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame SONDA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Weibull SONDA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2: # weibull_min.fit precisa de pelo menos 2 pontos
        print(f"⚠️ AVISO: A série SONDA '{column}' está vazia ou tem poucos dados após remover NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Weibull SONDA não gerado: Série '{column}' vazia ou com dados insuficientes após dropna.")
        print('-' * 100)
        return

    # Ajustar a distribuição de Weibull aos dados
    # floc=0 fixa o parâmetro de localização em 0, comum para velocidade do vento (não pode ser negativa)
    try:
        shape, loc, scale = weibull_min.fit(data_series, floc=0)
    except Exception as e:
        print(f"❌ ERRO ao ajustar a distribuição de Weibull para a coluna '{column}': {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}': {str(e)}")
        print('-' * 100)
        return

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data_series, fill=True, label='Densidade Empírica', color=color_density, linewidth=2)

    # Gerar pontos para a curva Weibull teórica
    x_weibull = np.linspace(data_series.min(), data_series.max(), 200)
    y_weibull = weibull_min.pdf(x_weibull, shape, loc, scale)
    plt.plot(x_weibull, y_weibull, color_weibull, linestyle='--', linewidth=2, label='Distribuição Weibull Teórica')

    # Adicionar linha vertical para a média (opcional, mas pode ser útil para referência)
    media = data_series.mean()
    plt.axvline(media, color='black', linestyle=':', linewidth=1.5, label=f'Média Empírica: {media:.2f}')

    plt.annotate(
        f'Weibull Parâmetros:\nShape (k): {shape:.2f}\nScale (λ): {scale:.2f}\nLoc (γ): {loc:.2f}', # loc será 0 devido a floc=0
        xy=(0.95, 0.85),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Densidade Weibull para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Carregamento do DataFrame dfSonda_jan2006_abr2007 ---
# (Mesma lógica de carregamento e pré-processamento da célula anterior)
print('=' * 100)
print("--- Carregando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan)
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')
    print("☑ DataFrame carregado e colunas relevantes convertidas para numérico.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gráficos.")

except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfSonda_jan2006_abr2007: {str(e)}. Gráficos SONDA não gerados.")
    logging.error(f"Erro no setup para gráficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gráfico de densidade com Weibull para ws_50 (SONDA)
    plot_density_with_weibull_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50', # Foco no atributo ws_50
        'Distribuição de Densidade (SONDA) com Ajuste Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_weibull_SondaWS50_melhorado.png',
        color_density='mediumseagreen', # Cor alterada para exemplo
        color_weibull='darkorange'      # Cor alterada para exemplo
    )
else:
    print("Skipping SONDA density plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Densidade com Weibull para Rede SONDA Concluída. ---")
print("-" * 100)
````

📚 **Distribuição de Weibull:**

A distribuição de Weibull é uma distribuição de probabilidade contínua que se destaca por sua versatilidade. Ela é definida por dois ou três parâmetros: o parâmetro de forma **`(k)`**, o parâmetro de escala **`(λ ou c)`**, e, opcionalmente, um parâmetro de localização **`(γ ou θ)`**.

- **Parâmetro de Forma (k):** Determina a forma fundamental da função de densidade de probabilidade. Valores diferentes de k podem fazer com que a distribuição de Weibull se assemelhe a outras distribuições. Por exemplo:
 - **`Se k < 1`**, a taxa de falha (ou, analogamente, a probabilidade de ocorrência de certas velocidades de vento) diminui com o tempo (ou velocidade).
 - **`Se k = 1`**, a distribuição de Weibull se reduz a uma distribuição exponencial, indicando uma taxa de falha constante.
 - **`Se k = 2`**, ela se torna uma distribuição de Rayleigh, comum para descrever a distribuição da velocidade do vento quando a direção do vento é uniforme.
 - **`Se k está entre 3 e 4`**, a distribuição de Weibull se aproxima da forma de uma distribuição normal.

- **Parâmetro de Escala (λ):** Este parâmetro "estica" ou "comprime" a distribuição ao longo do eixo horizontal (neste caso, o eixo da velocidade do vento). Ele está diretamente relacionado à velocidade média do vento; um valor maior de λ geralmente implica velocidades médias de vento mais altas.

- **Parâmetro de Localização (γ):** Define o ponto de início da distribuição. Para a velocidade do vento, este parâmetro é frequentemente fixado em zero, pois a velocidade não pode ser negativa.

 💡 A **`distribuição de Weibull`** é frequentemente empregada na análise de recursos eólicos, pois se ajusta bem a uma ampla gama de dados de velocidade do vento observados em diferentes locais. Sua flexibilidade permite modelar distribuições que podem ser simétricas ou assimétricas, o que é crucial para dados de vento que raramente seguem uma distribuição normal perfeita. Conforme afirma a literatura especializada, "a função de densidade de probabilidade de Weibull de dois parâmetros tem sido amplamente utilizada na indústria de energia eólica para descrever as distribuições de velocidade do vento".

#### 📈 **Gráfico de Densidade com curva Normal e Weibull**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado
from scipy.stats import norm, weibull_min # Para as curvas de distribuição

# Configurar logging (garanta que só é configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Densidade Combinado ---
def plot_density_combined_sonda(dataframe, column, title_prefix, xlabel_text, output_filename,
                                color_density='skyblue', color_normal='red', color_weibull='green'):
    """
    Gera um gráfico de densidade para uma coluna específica do DataFrame da Rede SONDA,
    incluindo curvas teóricas Normal e Weibull, e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        column (str): O nome da coluna para gerar o gráfico de densidade.
        title_prefix (str): Prefixo para o título do gráfico.
        xlabel_text (str): Rótulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        color_density (str, optional): Cor da densidade empírica.
        color_normal (str, optional): Cor da curva normal.
        color_weibull (str, optional): Cor da curva Weibull.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Densidade Combinado para SONDA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame SONDA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Combinado SONDA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return None, None # Retorna None para os parâmetros

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2:
        print(f"⚠️ AVISO: A série SONDA '{column}' está vazia ou tem poucos dados após remover NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Combinado SONDA não gerado: Série '{column}' vazia ou com dados insuficientes.")
        print('-' * 100)
        return None, None # Retorna None para os parâmetros

    # Parâmetros para a curva Normal
    media_normal = data_series.mean()
    desvio_normal = data_series.std()
    params_normal = {'media': media_normal, 'desvio': desvio_normal}

    # Ajustar a distribuição de Weibull
    try:
        shape_weibull, loc_weibull, scale_weibull = weibull_min.fit(data_series, floc=0)
        params_weibull = {'shape': shape_weibull, 'loc': loc_weibull, 'scale': scale_weibull}
    except Exception as e:
        print(f"❌ ERRO ao ajustar a distribuição de Weibull para a coluna '{column}': {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}': {str(e)}")
        params_weibull = None # Indica que o ajuste Weibull falhou
        print('-' * 100)
        # Continuamos para plotar pelo menos a Normal, se possível

    plt.figure(figsize=(9, 6)) # Ajustado para melhor visualização com mais elementos
    sns.kdeplot(data_series, fill=True, label='Densidade Empírica', color=color_density, linewidth=2, alpha=0.6)

    # Gerar pontos para as curvas teóricas
    x_vals = np.linspace(data_series.min(), data_series.max(), 200)

    # Curva Normal
    y_normal = norm.pdf(x_vals, media_normal, desvio_normal)
    plt.plot(x_vals, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribuição Normal Teórica')

    # Curva Weibull (se o ajuste foi bem-sucedido)
    if params_weibull:
        y_weibull = weibull_min.pdf(x_vals, shape_weibull, loc_weibull, scale_weibull)
        plt.plot(x_vals, y_weibull, color_weibull, linestyle=':', linewidth=2, label='Distribuição Weibull Teórica')

    # Linha vertical para a média empírica
    plt.axvline(media_normal, color='black', linestyle='-.', linewidth=1, label=f'Média Empírica: {media_normal:.2f}')

    # Anotações
    annotation_text = f'Normal:\nMédia: {media_normal:.2f}\nDesvio Padrão: {desvio_normal:.2f}\n'
    if params_weibull:
        annotation_text += f'\nWeibull:\nShape (k): {shape_weibull:.2f}\nScale (λ): {scale_weibull:.2f}\nLoc (γ): {loc_weibull:.2f}'
    else:
        annotation_text += "\nWeibull: Ajuste falhou"

    plt.annotate(
        annotation_text,
        xy=(0.95, 0.95),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.9),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend(loc='upper left', bbox_to_anchor=(0.01, 0.95)) # Ajuste da posição da legenda
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Densidade Combinado para SONDA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)
    return params_normal, params_weibull


# --- Carregamento do DataFrame dfSonda_jan2006_abr2007 ---
# (Mesma lógica de carregamento e pré-processamento da célula anterior)
print('=' * 100)
print("--- Carregando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan)
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')
    print("☑ DataFrame carregado e colunas relevantes convertidas para numérico.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gráficos.")

except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfSonda_jan2006_abr2007: {str(e)}. Gráficos SONDA não gerados.")
    logging.error(f"Erro no setup para gráficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gráfico de densidade combinado para ws_50 (SONDA)
    params_normal_ws50, params_weibull_ws50 = plot_density_combined_sonda(
        dfSonda_jan2006_abr2007,
        'ws_50', # Foco no atributo ws_50
        'Densidade - SONDAWS50: Empírica vs Normal vs Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_Normal_Weibull_SondaWS50_melhorado.png'
    )

    # Exibir parâmetros estimados se foram calculados
    if params_normal_ws50:
        print(f"\n📊 Parâmetros Estimados para Distribuição Normal (ws_50):")
        print(f"   - Média: {params_normal_ws50['media']:.2f}")
        print(f"   - Desvio Padrão: {params_normal_ws50['desvio']:.2f}")
    if params_weibull_ws50:
        print(f"\n📈 Parâmetros Estimados para Distribuição Weibull (ws_50):")
        print(f"   - Shape (k): {params_weibull_ws50['shape']:.2f}")
        print(f"   - Scale (λ): {params_weibull_ws50['scale']:.2f}")
        print(f"   - Loc (γ): {params_weibull_ws50['loc']:.2f}") # Será 0 se floc=0 foi usado
else:
    print("Skipping SONDA density plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Densidade Combinados para Rede SONDA Concluída. ---")
print("-" * 100)
````

#### 📈 **Gráfico de Dispersão Temporal - Velocidade do Vento a 50m**

````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado

# Configurar logging (garanta que só é configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Dispersão Temporal ---
def plot_temporal_dispersion_sonda(dataframe, time_column, data_column, title_prefix, xlabel_text, ylabel_text, output_filename,
                                   raw_data_color='skyblue', trend_color='red', raw_data_label_suffix='(Dados Horários)'):
    """
    Gera um gráfico de dispersão temporal para uma coluna específica do DataFrame da Rede SONDA,
    incluindo uma linha de tendência LOESS e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da Rede SONDA.
        time_column (str): Nome da coluna de data/hora.
        data_column (str): Nome da coluna de dados para plotar.
        title_prefix (str): Prefixo para o título do gráfico.
        xlabel_text (str): Rótulo do eixo X.
        ylabel_text (str): Rótulo do eixo Y.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        raw_data_color (str, optional): Cor dos dados brutos.
        trend_color (str, optional): Cor da linha de tendência.
        raw_data_label_suffix (str, optional): Sufixo para a legenda dos dados brutos.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Dispersão Temporal para SONDA: {data_column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or time_column not in dataframe.columns or data_column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame SONDA está vazio ou as colunas '{time_column}' ou '{data_column}' não existem. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Dispersão SONDA não gerado: DataFrame vazio ou colunas ausentes.")
        print('-' * 100)
        return

    # Certificar que a coluna de tempo é do tipo datetime
    try:
        df_plot = dataframe.copy() # Trabalhar com uma cópia para evitar SettingWithCopyWarning
        df_plot[time_column] = pd.to_datetime(df_plot[time_column])
    except Exception as e:
        print(f"❌ ERRO ao converter a coluna de tempo '{time_column}' para datetime: {str(e)}")
        logging.error(f"Erro ao converter '{time_column}' para datetime: {str(e)}")
        print('-' * 100)
        return

    data_series = df_plot[data_column].dropna()
    time_series = df_plot.loc[data_series.index, time_column] # Alinhar tempo com dados não nulos

    if data_series.empty:
        print(f"⚠️ AVISO: A série SONDA '{data_column}' está vazia após remover NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Dispersão SONDA não gerado: Série '{data_column}' vazia após dropna.")
        print('-' * 100)
        return

    plt.figure(figsize=(12, 5)) # Ajustado para melhor visualização temporal

    # Plotar os dados brutos (pontos menores e com transparência devido à densidade)
    # Usar plt.plot para dados de alta frequência pode ser mais performático e visualmente limpo
    plt.plot(time_series, data_series, label=f'{data_column} {raw_data_label_suffix}', color=raw_data_color, alpha=0.6, linestyle='-', marker=None, linewidth=0.8)

    # Linha de tendência LOESS
    # regplot precisa de valores numéricos para x se não interpretar datetime diretamente
    # Convertendo para números de data matplotlib se necessário, embora seaborn mais recente lide melhor com datetime
    time_numeric = mdates.date2num(time_series)
    sns.regplot(
        x=time_numeric,
        y=data_series,
        scatter=False, # Já plotamos os dados brutos
        lowess=True,   # Usa LOESS para uma tendência mais suave e não linear
        color=trend_color,
        label='Tendência (LOESS)',
        line_kws={'linewidth': 2}
    )

    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=12))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45, ha='right')

    plt.title(f'{title_prefix} - {data_column} (01/2006 a 04/2007)')
    plt.xlabel(xlabel_text)
    plt.ylabel(ylabel_text)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Dispersão Temporal para SONDA '{data_column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Carregamento e Pré-processamento do DataFrame dfSonda_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando e Pré-processando DataFrame dfSonda_jan2006_abr2007 (Rede SONDA) ---")
print('=' * 100)
try:
    caminho_df_sonda_jan2006_abr2007_final = os.path.join(
        '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/DadosSonda_Unificados_Jan2006Apr2007',
        'dfSonda_jan2006_abr2007.csv'
    )
    dfSonda_jan2006_abr2007 = pd.read_csv(caminho_df_sonda_jan2006_abr2007_final, sep=";", encoding="latin1")

    # Converter 'datetm' para datetime, crucial para plots temporais
    dfSonda_jan2006_abr2007['datetm'] = pd.to_datetime(dfSonda_jan2006_abr2007['datetm'], errors='coerce')

    # Garantir conversão de N/A e N/S para NaN e tipo float para colunas de medição
    cols_to_convert = ['ws_25', 'wd_25', 'tp_25', 'ws_50', 'wd_50', 'tp_50']
    for col in cols_to_convert:
        if col in dfSonda_jan2006_abr2007.columns:
            dfSonda_jan2006_abr2007[col] = dfSonda_jan2006_abr2007[col].replace(['N/A', 'N/S', -999, -999.0], np.nan)
            dfSonda_jan2006_abr2007[col] = pd.to_numeric(dfSonda_jan2006_abr2007[col], errors='coerce')

    # Remover linhas onde 'datetm' ou 'ws_50' são NaT/NaN após conversão, para evitar problemas no plot
    dfSonda_jan2006_abr2007.dropna(subset=['datetm', 'ws_50'], inplace=True)

    print("☑ DataFrame carregado e colunas relevantes convertidas/limpas.")
    logging.info("DataFrame dfSonda_jan2006_abr2007 carregado e processado para gráficos.")

except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_sonda_jan2006_abr2007_final)}' não encontrado.")
    dfSonda_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfSonda_jan2006_abr2007: {str(e)}. Gráficos SONDA não gerados.")
    logging.error(f"Erro no setup para gráficos SONDA: {str(e)}")
    dfSonda_jan2006_abr2007 = pd.DataFrame()

if not dfSonda_jan2006_abr2007.empty:
    # Gerar gráfico de dispersão temporal para ws_50 (SONDA)
    plot_temporal_dispersion_sonda(
        dfSonda_jan2006_abr2007,
        time_column='datetm',       # Coluna de tempo
        data_column='ws_50',        # Coluna de dados
        title_prefix='Dispersão Temporal (SONDA)',
        xlabel_text='Tempo',
        ylabel_text='Velocidade do Vento (m/s) a 50m',
        output_filename='./dispersao_temporal_SondaWS50_melhorado.png',
        raw_data_label_suffix='(Dados 10 min)' # Ajustado para refletir a frequência original
    )
else:
    print("Skipping SONDA dispersion plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Dispersão Temporal para Rede SONDA Concluída. ---")
print("-" * 100)
````

---
📚 **O que é a Curva LOESS?**

>LOESS é uma técnica de regressão não paramétrica robusta que visa criar uma curva suave através de um gráfico de dispersão de pontos. Diferentemente de uma regressão linear simples, que ajusta uma única reta a todos os dados, o LOESS ajusta múltiplos modelos de regressão simples a subconjuntos localizados dos dados. Isso o torna particularmente eficaz para revelar tendências e ciclos em dados que podem não seguir um padrão linear global ou uma função matemática específica.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado

# --- Função Auxiliar (Repetida para Independência) ---
def plot_boxplot_with_outliers_info_nasa(dataframe, column, title_prefix, xlabel_text, output_filename, color='skyblue'):
    """
    Gera um boxplot para uma coluna específica do DataFrame da NASA POWER,
    incluindo informações de outliers (IQR) e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da NASA.
        column (str): O nome da coluna para gerar o boxplot.
        title_prefix (str): Prefixo para o título do gráfico (ex: "Boxplot -").
        xlabel_text (str): Rótulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        color (str, optional): Cor do boxplot. Padrão é 'skyblue'.
    """
    print('=' * 100)
    print(f"--- Gerando Boxplot para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame NASA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o boxplot.")
        logging.warning(f"Boxplot NASA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"⚠️ AVISO: A série NASA '{column}' está vazia após remover NAs. Não foi possível gerar o boxplot.")
        logging.warning(f"Boxplot NASA não gerado: Série '{column}' vazia após dropna.")
        print('-' * 100)
        return

    Q1 = data_series.quantile(0.25)
    Q3 = data_series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound_iqr = Q1 - 1.5 * IQR
    upper_bound_iqr = Q3 + 1.5 * IQR

    outliers_count = data_series[(data_series < lower_bound_iqr) | (data_series > upper_bound_iqr)].count()
    total_count = len(data_series)
    outliers_percentage = (outliers_count / total_count * 100) if total_count > 0 else 0

    plt.figure(figsize=(9, 3))
    sns.boxplot(x=data_series, color=color, linewidth=1.5, flierprops=dict(marker='o', markerfacecolor='red', markersize=5))

    plt.annotate(
        f'Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\nOutliers: {outliers_count} ({outliers_percentage:.2f}%)',
        xy=(0.95, 0.05),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='bottom'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Boxplot para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)


# --- Aplicação para dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Gerando gráficos para dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)

# O DataFrame dfNASA_jan2006_abr2007 deve estar carregado nas células anteriores.
# Exemplo de como garantir que ele esteja carregado, se esta célula for executada isoladamente:
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
except Exception as e:
    print(f"❌ ERRO ao carregar dfNASA_jan2006_abr2007: {str(e)}. Gráficos NASA não gerados.")
    logging.error(f"Erro no setup para gráficos NASA: {str(e)}")
    dfNASA_jan2006_abr2007 = pd.DataFrame() # Define como vazio para pular gráficos

if not dfNASA_jan2006_abr2007.empty:
    # Gerar boxplot para WS50M (NASA)
    plot_boxplot_with_outliers_info_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M',
        'Boxplot - Velocidade do Vento 50m (NASA)',
        'Velocidade (m/s)',
        './boxplot_NASAWS50.png',
        color='skyblue'
    )

    # Gerar boxplot para WD50M (NASA)
    plot_boxplot_with_outliers_info_nasa(
        dfNASA_jan2006_abr2007,
        'WD50M',
        'Boxplot - Direção do Vento 50m (NASA)',
        'Direção (°)',
        './boxplot_NASAWD50.png',
        color='lightgreen'
    )
else:
    print("Skipping NASA boxplots due to empty or problematic DataFrame.")

print("--- Geração de Boxplots para NASA POWER Concluída. ---")
print("-" * 100)
````

#### 📈 **Gráfico de Densidade com curva normal teórica**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd # Garanta que pandas está importado
import logging # Garanta que logging está importado
from IPython.display import display # Garanta que display está importado
from scipy.stats import norm # Para a curva de distribuição normal

# Configurar logging (garanta que só é configurado uma vez)
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Densidade (NASA) ---
def plot_density_with_normal_nasa(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='lightcoral', color_normal='darkred'):
    """
    Gera um gráfico de densidade para uma coluna específica do DataFrame da NASA,
    incluindo uma curva de distribuição normal teórica e salva o gráfico.

    Args:
        dataframe (pd.DataFrame): O DataFrame da NASA.
        column (str): O nome da coluna para gerar o gráfico de densidade.
        title_prefix (str): Prefixo para o título do gráfico.
        xlabel_text (str): Rótulo do eixo X.
        output_filename (str): Nome do arquivo para salvar o gráfico.
        color_density (str, optional): Cor da densidade empírica.
        color_normal (str, optional): Cor da curva normal.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Densidade para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame NASA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade NASA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty:
        print(f"⚠️ AVISO: A série NASA '{column}' está vazia após remover NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade NASA não gerado: Série '{column}' vazia após dropna.")
        print('-' * 100)
        return

    media = data_series.mean()
    desvio = data_series.std()

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data_series, fill=True, label='Densidade Empírica', color=color_density, linewidth=2, alpha=0.7)

    x_normal = np.linspace(data_series.min(), data_series.max(), 200)
    y_normal = norm.pdf(x_normal, media, desvio)
    plt.plot(x_normal, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribuição Normal Teórica')

    plt.axvline(media, color='black', linestyle=':', linewidth=1.5, label=f'Média: {media:.2f}')

    plt.annotate(
        f'Média: {media:.2f}\nDesvio Padrão: {desvio:.2f}',
        xy=(0.95, 0.85),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Densidade para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)

    # Certificar que a coluna WS50M é numérica (já deve ser, mas é uma boa prática)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns:
        dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')

    print("☑ DataFrame NASA carregado e coluna WS50M verificada/convertida para numérico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gráficos.")

except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' não encontrado.")
    logging.error(f"Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' não encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfNASA_jan2006_abr2007: {str(e)}. Gráficos NASA não gerados.")
    logging.error(f"Erro no setup para gráficos NASA: {str(e)}")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    plot_density_with_normal_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M', # Coluna da NASA para velocidade do vento a 50m
        'Distribuição de Densidade (NASA)',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_normal_NASAWS50_melhorado.png'
    )
else:
    print("Skipping NASA density plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Densidade com Normal para NASA POWER Concluída. ---")
print("-" * 100)
````

#### 📈 **Gráfico de Distribuição de Densidade - NASAWS50 (Weibull)**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd
import logging
from IPython.display import display
from scipy.stats import weibull_min # Para a curva de distribuição Weibull

# Configurar logging
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Densidade com Weibull (NASA) ---
def plot_density_with_weibull_nasa(dataframe, column, title_prefix, xlabel_text, output_filename, color_density='lightgreen', color_weibull='darkgreen'):
    """
    Gera um gráfico de densidade para uma coluna específica do DataFrame da NASA,
    incluindo uma curva de distribuição Weibull teórica e salva o gráfico.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Densidade com Weibull para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame NASA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Weibull NASA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2:
        print(f"⚠️ AVISO: A série NASA '{column}' está vazia ou tem poucos dados após remover NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Weibull NASA não gerado: Série '{column}' vazia ou com dados insuficientes.")
        print('-' * 100)
        return

    try:
        shape, loc, scale = weibull_min.fit(data_series, floc=0)
    except Exception as e:
        print(f"❌ ERRO ao ajustar a distribuição de Weibull para a coluna '{column}': {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}' (NASA): {str(e)}")
        print('-' * 100)
        return

    plt.figure(figsize=(8, 5))
    sns.kdeplot(data_series, fill=True, label='Densidade Empírica', color=color_density, linewidth=2, alpha=0.7)

    x_weibull = np.linspace(data_series.min(), data_series.max(), 200)
    y_weibull = weibull_min.pdf(x_weibull, shape, loc, scale)
    plt.plot(x_weibull, y_weibull, color_weibull, linestyle='--', linewidth=2, label='Distribuição Weibull Teórica')

    media_empirica = data_series.mean()
    plt.axvline(media_empirica, color='black', linestyle=':', linewidth=1.5, label=f'Média Empírica: {media_empirica:.2f}')

    plt.annotate(
        f'Weibull Parâmetros:\nShape (k): {shape:.2f}\nScale (λ): {scale:.2f}\nLoc (γ): {loc:.2f}',
        xy=(0.95, 0.85),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.8),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Densidade Weibull para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns:
        dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')
    print("☑ DataFrame NASA carregado e coluna WS50M verificada/convertida para numérico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gráficos.")
except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' não encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfNASA_jan2006_abr2007: {str(e)}. Gráficos NASA não gerados.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    plot_density_with_weibull_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M',
        'Distribuição de Densidade (NASA) com Ajuste Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_weibull_NASAWS50_melhorado.png'
    )
else:
    print("Skipping NASA density plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Densidade com Weibull para NASA POWER Concluída. ---")
print("-" * 100)
````

#### 📈 **Gráfico de Densidade com curva Normal e Weibull**

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import pandas as pd
import logging
from IPython.display import display
from scipy.stats import norm, weibull_min

# Configurar logging
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Densidade Combinado (NASA) ---
def plot_density_combined_nasa(dataframe, column, title_prefix, xlabel_text, output_filename,
                               color_density='palegreen', color_normal='darkred', color_weibull='darkorange'):
    """
    Gera um gráfico de densidade para uma coluna específica do DataFrame da NASA,
    incluindo curvas teóricas Normal e Weibull, e salva o gráfico.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Densidade Combinado para NASA: {column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or column not in dataframe.columns:
        print(f"⚠️ AVISO: DataFrame NASA está vazio ou a coluna '{column}' não existe. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Combinado NASA não gerado: DataFrame vazio ou coluna '{column}' ausente.")
        print('-' * 100)
        return None, None

    data_series = dataframe[column].dropna()

    if data_series.empty or len(data_series) < 2:
        print(f"⚠️ AVISO: A série NASA '{column}' está vazia ou tem poucos dados. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Densidade Combinado NASA não gerado: Série '{column}' vazia ou com dados insuficientes.")
        print('-' * 100)
        return None, None

    media_normal = data_series.mean()
    desvio_normal = data_series.std()
    params_normal = {'media': media_normal, 'desvio': desvio_normal}

    try:
        shape_weibull, loc_weibull, scale_weibull = weibull_min.fit(data_series, floc=0)
        params_weibull = {'shape': shape_weibull, 'loc': loc_weibull, 'scale': scale_weibull}
    except Exception as e:
        print(f"❌ ERRO ao ajustar a distribuição de Weibull para '{column}' (NASA): {str(e)}")
        logging.error(f"Erro ao ajustar Weibull para '{column}' (NASA): {str(e)}")
        params_weibull = None
        print('-' * 100)

    plt.figure(figsize=(9, 6))
    sns.kdeplot(data_series, fill=True, label='Densidade Empírica', color=color_density, linewidth=2, alpha=0.6)

    x_vals = np.linspace(data_series.min(), data_series.max(), 200)

    y_normal = norm.pdf(x_vals, media_normal, desvio_normal)
    plt.plot(x_vals, y_normal, color_normal, linestyle='--', linewidth=2, label='Distribuição Normal Teórica')

    if params_weibull:
        y_weibull = weibull_min.pdf(x_vals, shape_weibull, loc_weibull, scale_weibull)
        plt.plot(x_vals, y_weibull, color_weibull, linestyle=':', linewidth=2, label='Distribuição Weibull Teórica')

    plt.axvline(media_normal, color='black', linestyle='-.', linewidth=1, label=f'Média Empírica: {media_normal:.2f}')

    annotation_text = f'Normal:\nMédia: {media_normal:.2f}\nDesvio Padrão: {desvio_normal:.2f}\n'
    if params_weibull:
        annotation_text += f'\nWeibull:\nShape (k): {shape_weibull:.2f}\nScale (λ): {scale_weibull:.2f}\nLoc (γ): {loc_weibull:.2f}'
    else:
        annotation_text += "\nWeibull: Ajuste falhou"

    plt.annotate(
        annotation_text,
        xy=(0.95, 0.95),
        xycoords='axes fraction',
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", lw=0.5, alpha=0.9),
        ha='right',
        va='top'
    )

    plt.title(f'{title_prefix} - {column}')
    plt.xlabel(xlabel_text)
    plt.ylabel('Densidade')
    plt.legend(loc='upper left', bbox_to_anchor=(0.01, 0.95))
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Densidade Combinado para NASA '{column}' salvo em '{output_filename}'.")
    print('-' * 100)
    return params_normal, params_weibull

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns:
        dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')
    print("☑ DataFrame NASA carregado e coluna WS50M verificada/convertida para numérico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gráficos.")
except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' não encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfNASA_jan2006_abr2007: {str(e)}. Gráficos NASA não gerados.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    params_normal_nasa_ws50, params_weibull_nasa_ws50 = plot_density_combined_nasa(
        dfNASA_jan2006_abr2007,
        'WS50M',
        'Densidade - NASA WS50M: Empírica vs Normal vs Weibull',
        'Velocidade do Vento (m/s) a 50m',
        './densidade_Normal_Weibull_NASAWS50_melhorado.png'
    )

    if params_normal_nasa_ws50:
        print(f"\n📊 Parâmetros Estimados para Distribuição Normal (WS50M - NASA):")
        print(f"   - Média: {params_normal_nasa_ws50['media']:.2f}")
        print(f"   - Desvio Padrão: {params_normal_nasa_ws50['desvio']:.2f}")
    if params_weibull_nasa_ws50:
        print(f"\n📈 Parâmetros Estimados para Distribuição Weibull (WS50M - NASA):")
        print(f"   - Shape (k): {params_weibull_nasa_ws50['shape']:.2f}")
        print(f"   - Scale (λ): {params_weibull_nasa_ws50['scale']:.2f}")
        print(f"   - Loc (γ): {params_weibull_nasa_ws50['loc']:.2f}")
else:
    print("Skipping NASA combined density plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Densidade Combinados para NASA POWER Concluída. ---")
print("-" * 100)
````

````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import os
import pandas as pd
import logging
from IPython.display import display

# Configurar logging
if not logging.getLogger().hasHandlers():
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# --- Função Auxiliar Melhorada para Gráfico de Dispersão Temporal (NASA) ---
def plot_temporal_dispersion_nasa(dataframe, year_col, mo_col, dy_col, hr_col, data_column,
                                  title_prefix, xlabel_text, ylabel_text, output_filename,
                                  raw_data_color='dodgerblue', trend_color='darkorange', raw_data_label_suffix='(Dados Horários)'):
    """
    Gera um gráfico de dispersão temporal para uma coluna específica do DataFrame da NASA,
    incluindo uma linha de tendência LOESS e salva o gráfico.
    """
    print('=' * 100)
    print(f"--- Gerando Gráfico de Dispersão Temporal para NASA: {data_column} ({title_prefix}) ---")
    print('=' * 100)

    if dataframe.empty or not all(col in dataframe.columns for col in [year_col, mo_col, dy_col, hr_col, data_column]):
        print(f"⚠️ AVISO: DataFrame NASA está vazio ou colunas de tempo/dados não existem. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Dispersão NASA não gerado: DataFrame vazio ou colunas ausentes.")
        print('-' * 100)
        return

    df_plot = dataframe.copy()

    # Criar coluna datetime
    try:
        df_plot['datetm'] = pd.to_datetime(df_plot[year_col].astype(str) + '-' +
                                           df_plot[mo_col].astype(str) + '-' +
                                           df_plot[dy_col].astype(str) + ' ' +
                                           df_plot[hr_col].astype(str).str.zfill(2) + ':00:00', errors='coerce')
    except Exception as e:
        print(f"❌ ERRO ao criar a coluna 'datetm' a partir das colunas de data/hora: {str(e)}")
        logging.error(f"Erro ao criar 'datetm' para NASA: {str(e)}")
        print('-' * 100)
        return

    df_plot.dropna(subset=['datetm', data_column], inplace=True) # Remove linhas onde datetm ou data_column são NaT/NaN

    if df_plot.empty:
        print(f"⚠️ AVISO: DataFrame NASA ficou vazio após criação de 'datetm' e remoção de NAs. Não foi possível gerar o gráfico.")
        logging.warning(f"Gráfico de Dispersão NASA não gerado: DataFrame vazio após processamento de 'datetm'.")
        print('-' * 100)
        return

    data_series = df_plot[data_column]
    time_series = df_plot['datetm']

    plt.figure(figsize=(12, 5))
    plt.plot(time_series, data_series, label=f'{data_column} {raw_data_label_suffix}', color=raw_data_color, alpha=0.5, linestyle='-', marker=None, linewidth=0.8)

    time_numeric = mdates.date2num(time_series)
    sns.regplot(
        x=time_numeric,
        y=data_series,
        scatter=False,
        lowess=True,
        color=trend_color,
        label='Tendência (LOESS)',
        line_kws={'linewidth': 2}
    )

    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(minticks=5, maxticks=12))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.xticks(rotation=45, ha='right')

    plt.title(f'{title_prefix} - {data_column} (01/2006 a 04/2007)')
    plt.xlabel(xlabel_text)
    plt.ylabel(ylabel_text)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(output_filename, dpi=300)
    print(f"☑ Gráfico salvo em: '{output_filename}'")
    plt.show()
    logging.info(f"Gráfico de Dispersão Temporal para NASA '{data_column}' salvo em '{output_filename}'.")
    print('-' * 100)

# --- Carregamento do DataFrame dfNASA_jan2006_abr2007 ---
print('=' * 100)
print("--- Carregando DataFrame dfNASA_jan2006_abr2007 (NASA POWER) ---")
print('=' * 100)
try:
    caminho_df_nasa_jan2006_abr2007 = '/content/drive/MyDrive/Disciplinas/Ciência de Dados/Projeto de Disciplina/dfNASA_jan2006_abr2007.csv'
    dfNASA_jan2006_abr2007 = pd.read_csv(caminho_df_nasa_jan2006_abr2007)
    if 'WS50M' in dfNASA_jan2006_abr2007.columns: # Necessário para o data_column
         dfNASA_jan2006_abr2007['WS50M'] = pd.to_numeric(dfNASA_jan2006_abr2007['WS50M'], errors='coerce')
    print("☑ DataFrame NASA carregado e coluna WS50M verificada/convertida para numérico.")
    logging.info("DataFrame dfNASA_jan2006_abr2007 carregado para gráficos.")
except FileNotFoundError:
    print(f"❌ ERRO: Arquivo '{os.path.basename(caminho_df_nasa_jan2006_abr2007)}' não encontrado.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()
except Exception as e:
    print(f"❌ ERRO ao carregar ou pré-processar dfNASA_jan2006_abr2007: {str(e)}. Gráficos NASA não gerados.")
    dfNASA_jan2006_abr2007 = pd.DataFrame()

if not dfNASA_jan2006_abr2007.empty:
    plot_temporal_dispersion_nasa(
        dfNASA_jan2006_abr2007,
        year_col='YEAR', # Coluna de ano
        mo_col='MO',     # Coluna de mês
        dy_col='DY',     # Coluna de dia
        hr_col='HR',     # Coluna de hora
        data_column='WS50M',
        title_prefix='Dispersão Temporal (NASA)',
        xlabel_text='Tempo',
        ylabel_text='Velocidade do Vento (m/s) a 50m',
        output_filename='./dispersao_temporal_NASAWS50_melhorado.png'
    )
else:
    print("Skipping NASA dispersion plots due to empty or problematic DataFrame.")

print("--- Geração de Gráficos de Dispersão Temporal para NASA POWER Concluída. ---")
print("-" * 100)
````

### 🔨 **3.1. Transformar Dataframe da REDE SONDA (01/2006-04/2007)**

Os dados da REDE SONDA, originalmente registrados a cada 10 minutos, foram reamostrados para frequência **horária**. Para tornar os dados mais robustos:

- Foi realizada a **conversão da coluna de data/hora** (`datetm`) para o tipo `datetime`.
- Aplicada **reamostragem horária** com dois métodos de agregação:
  - **Média** (`mean`)
  - **Mediana** (`median`)
- Gerada uma **média suavizada**, combinando média e mediana:
  
$$
\text{Média Suavizada} = \frac{\text{Média} + \text{Mediana}}{2}
$$

Essa abordagem reduz o impacto de outliers e flutuações abruptas, proporcionando uma melhor estimativa do comportamento horário dos dados de vento.

### 🧮 **Cálculo do Erro Padrão entre SONDAWS50 e NASAWS50**
 O erro padrão da estimativa (SEE) mede a dispersão média dos dados em torno da linha de regressão.
  
 **Fórmula**:  

 $$
 \text{SEE} = \sqrt{\frac{\sum (y_i - \hat{y}_i)^2}{n - 2}}
 $$  

 Onde:  
 > - Valores reais de `NASAWS50`:<left>$$y_i$$<left>
 > - Valores previstos pela regressão linear: $$ \hat{y}_i $$  
 > - Número de observações: $$n$$

````python
from sklearn.linear_model import LinearRegression

# Supondo df_grouped já contém as médias suavizadas
X = df_grouped[['SONDAWS50_svz']]  # Variável independente
y = df_grouped['NASAWS50_svz']     # Variável dependente

# Ajustar modelo de regressão linear
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)  # Valores previstos

# Calcular resíduos e SEE
residuos = y - y_pred
soma_quadrados_residuos = np.sum(residuos**2)
n = len(y)
SEE = np.sqrt(soma_quadrados_residuos / (n - 2))

print(f'Erro Padrão da Estimativa (SEE):\n{SEE:.3f} m/s')
````

#### ✅ **Análise de Resultados:**

- **`SEE = 0.563 m/s`**: Indica que, em média, os valores de NASAWS50 desviam-se **`±0.563m/s`** da linha de regressão ajustada.

- Quanto menor o **`SEE`**, mais precisa é a relação linear. Neste caso, o valor sugere uma dispersão moderada, consistente com o **`R²`** de **`0.47`**.

 O cálculo foi feito com médias mensais suavizadas, o que pode subestimar a variabilidade real. Para maior precisão, utilize os dados brutos.

2. **Comparação com Outras Métricas:**

 - O **`SEE`** complementa o **`R²`**: Um **`R²`** baixo e **`SEE`** alto reforçam a necessidade de investigar outros fatores explicativos.

 - No exemplo, **`47%`** da variância é explicada, e o **`SEE`** quantifica o erro médio absoluto.

**Download multiponto com Python: API NASA POWER DAV**

---
Esta é uma visão geral do processo para solicitar dados de vários locais de ponto único da API do POWER.

- A guia abaixo permite o download diretamente da API para qualquer formato oferecido. `Loop (Points)`

- O e suportam multiprocessamento e leem os dados da resposta JSON puxando os dados para um DataFrame do Pandas antes de serem exportados como CSV. Isso pode ser editado para criar qualquer formato com Python e/ou o módulo Pandas. `Multiprocessing (Points)Multiprocessing (Region)`


```
'''
*Version: 2.0 Published: 2021/03/09* Source: [NASA POWER](https://power.larc.nasa.gov/)
POWER API Multi-Point Download
This is an overview of the process to request data from multiple data points from the POWER API.
'''

import os, json, requests

locations = [(32.929, -95.770), (5, 10)]

output = r""
base_url = r"https://power.larc.nasa.gov/api/temporal/daily/point?parameters=T2M,T2MDEW,T2MWET,TS,T2M_RANGE,T2M_MAX,T2M_MIN&community=RE&longitude={longitude}&latitude={latitude}&start=20150101&end=20150305&format=JSON"

for latitude, longitude in locations:
    api_request_url = base_url.format(longitude=longitude, latitude=latitude)

    response = requests.get(url=api_request_url, verify=True, timeout=30.00)

    content = json.loads(response.content.decode('utf-8'))
    filename = response.headers['content-disposition'].split('filename=')[1]

    filepath = os.path.join(output, filename)
    with open(filepath, 'w') as file_object:
        json.dump(content, file_object)
```
